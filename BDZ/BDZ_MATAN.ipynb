```python
from matplotlib.pyplot import*
import numpy as np
import sympy as sp
x = sp.Symbol('x')
f = (x**3 + x**2 - x - 1)/(x**3 + x**2 - x + 1)
print(sp.limit(f,x,-1))
```

```python
import numpy as np
import matplotlib.pyplot as plt

#    
x = np.linspace(-3, 3, 1000)
y = (x**3 + x**2 - x - 1) / (x**3 - x**2 - x + 1)

#    (x = 1  x = -1)
mask = (np.abs(x - 1) > 0.01) & (np.abs(x + 1) > 0.01)
x_clean = x[mask]
y_clean = y[mask]

#  
plt.figure(figsize=(10, 6))
plt.plot(x_clean, y_clean, 'b-', linewidth=2, label=r'$f(x) = \frac{x^3 + x^2 - x - 1}{x^3 - x^2 - x + 1}$')

#   
plt.axvline(x=1, color='red', linestyle='--', alpha=0.7, label='  x=1')
plt.axvline(x=-1, color='orange', linestyle='--', alpha=0.7, label='  x=-1')

#    x  -1
plt.plot(-1, 0, 'go', markersize=8, label='  x-1: 0')

#   
plt.axhline(y=1, color='purple', linestyle='--', alpha=0.7, label='  y=1')

#  
plt.xlim(-2, 2)
plt.ylim(-5, 5)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('    ')
plt.grid(True, alpha=0.3)
plt.legend()

#  
plt.tight_layout()
plt.show()

```

```python
from matplotlib.pyplot import*
from numpy import *

def f(x):
    return x**2 / (np.cbrt(1 + 3*x) - 1 - x)


x = np.linspace(-1, 1, 2000)

mask = np.abs(x) > 0.001
x_clean = x[mask]
y_clean = f(x_clean)
xlim(-0.5,0.5)
ylim(-3,1)

axhline(-1,color='red',linestyle='--')
axvline(0,color='green',linestyle='--')
grid()

plot(x_clean, y_clean, 'b-', linewidth=2, 
         label=r'$f(x) = \frac{x^2}{\sqrt[3]{1+3x} - 1 - x}$')


plot(0, -1, 'ro', markersize=8, markeredgewidth=2, 
         markerfacecolor='red', label='  x0: -1')
```

```python
import numpy as np
import matplotlib.pyplot as plt
import math

#  
def f(x):
    return ((x**(1/3) + 2*x**(1/6) + 1) / (x**(1/3) + 3)) ** (x**(1/6))

#    
x = np.linspace(1, 10000, 1000)
y = f(x)

#  
plt.figure(figsize=(12, 8))

#  
plt.plot(x, y, 'b-', linewidth=2, 
         label=r'$f(x) = \left(\frac{\sqrt[3]{x} + 2\sqrt[6]{x} + 1}{\sqrt[3]{x} + 3}\right)^{\sqrt[6]{x}}$')

#  
limit_value = math.e**2
plt.axhline(y=limit_value, color='red', linestyle='--', alpha=0.7, 
           label=f': $e^2  {limit_value:.4f}$')

#  
plt.xlim(0, 10000)
plt.ylim(0, 10)
plt.xlabel('x', fontsize=12)
plt.ylabel('f(x)', fontsize=12)
plt.title('     x+', fontsize=14)
plt.grid(True, alpha=0.3)
plt.legend(fontsize=10)

plt.tight_layout()
plt.show()

#   -    x
plt.figure(figsize=(12, 8))

#      
x_log = np.logspace(0, 6, 1000)  #  10^0  10^6
y_log = f(x_log)

plt.semilogx(x_log, y_log, 'b-', linewidth=2)
plt.axhline(y=limit_value, color='red', linestyle='--', alpha=0.7, 
           label=f': $e^2  {limit_value:.4f}$')

plt.xlim(1, 10**6)
plt.ylim(0, 10)
plt.xlabel('x ( )', fontsize=12)
plt.ylabel('f(x)', fontsize=12)
plt.title('  (   x)', fontsize=14)
plt.grid(True, alpha=0.3)
plt.legend(fontsize=10)

plt.tight_layout()
plt.show()

#     
print("    x:")
test_x = [100, 1000, 10000, 100000, 1000000]
print(f": e = {limit_value:.6f}")
print("-" * 40)
for xi in test_x:
    print(f"f({xi}) = {f(xi):.6f}")
```
