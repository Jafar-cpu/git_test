```python
from matplotlib.pyplot import*
import numpy as np
import sympy as sp
x = sp.Symbol('x')
f = (x**3 + x**2 - x - 1)/(x**3 + x**2 - x + 1)
print(sp.limit(f,x,-1))
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Создаем данные для графика
x = np.linspace(-3, 3, 1000)
y = (x**3 + x**2 - x - 1) / (x**3 - x**2 - x + 1)

# Исключаем точки разрыва (x = 1 и x = -1)
mask = (np.abs(x - 1) > 0.01) & (np.abs(x + 1) > 0.01)
x_clean = x[mask]
y_clean = y[mask]

# Создаем график
plt.figure(figsize=(10, 6))
plt.plot(x_clean, y_clean, 'b-', linewidth=2, label=r'$f(x) = \frac{x^3 + x^2 - x - 1}{x^3 - x^2 - x + 1}$')

# Отмечаем особые точки
plt.axvline(x=1, color='red', linestyle='--', alpha=0.7, label='Вертикальная асимптота x=1')
plt.axvline(x=-1, color='orange', linestyle='--', alpha=0.7, label='Устранимая особенность x=-1')

# Отмечаем предел при x → -1
plt.plot(-1, 0, 'go', markersize=8, label='Предел при x→-1: 0')

# Отмечаем горизонтальную асимптоту
plt.axhline(y=1, color='purple', linestyle='--', alpha=0.7, label='Горизонтальная асимптота y=1')

# Настройки графика
plt.xlim(-2, 2)
plt.ylim(-5, 5)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('График функции и её пределы')
plt.grid(True, alpha=0.3)
plt.legend()

# Показываем график
plt.tight_layout()
plt.show()

```

```python
from matplotlib.pyplot import*
import numpy as np
import sympy as sp
x = sp.Symbol('x')
f = (x**2)/(sp.cbrt(1+3*x) -1 -x)
print(sp.limit(f,x,0))
```

```python
from matplotlib.pyplot import*
from numpy import *

def f(x):
    return x**2 / (np.cbrt(1 + 3*x) - 1 - x)


x = np.linspace(-1, 1, 2000)

mask = np.abs(x) > 0.001
x_clean = x[mask]
y_clean = f(x_clean)
xlim(-0.5,0.5)
ylim(-3,1)

axhline(-1,color='red',linestyle='--')
axvline(0,color='green',linestyle='--')
grid()

plot(x_clean, y_clean, 'b-', linewidth=2, 
         label=r'$f(x) = \frac{x^2}{\sqrt[3]{1+3x} - 1 - x}$')


plot(0, -1, 'ro', markersize=8, markeredgewidth=2, 
         markerfacecolor='red', label='Предел при x→0: -1')
```

```python
#3.116
from matplotlib.pyplot import*
from numpy import *

# Создаем данные для графика
x = np.linspace(-2, 2, 1000)
ylim(0,8)
xlim(-1.5,1.5)
axhline(y=5,color='red',linestyle='--',alpha=0.5)

axhline(y=0,color='red',linestyle='--',alpha=0.3)
axvline(x=0,color='black',linestyle='-',alpha=0.3)
# Исключаем точку x=0 чтобы избежать деления на 0
x_clean = x[x != 0]
def f(x):  
    return (1-cos(x) * (cos(2*x)))/(1-cos(x))

y_clean = f(x_clean)
grid()
plot(x_clean,y_clean,'b-',linewidth=1 )
plot(0,5,'ro',color='yellow')
show()
```

```python
#3.146

import numpy as np
import matplotlib.pyplot as plt
import math


def f(x):
    return ((x**(1/3) + 2*x**(1/6) + 1) / (x**(1/3) + 3)) ** (x**(1/6))


x = np.linspace(1, 10000, 1000)
y = f(x)
xlim(0,10000)
ylim(0,10)
limit = math.e**2
axhline(limit,color='red',linestyle='--')

plt.plot(x, y, 'b-', linewidth=2)

show()
```
