```python
from matplotlib.pyplot import*
import numpy as np
import sympy as sp
x = sp.Symbol('x')
f = (x**3 + x**2 - x - 1)/(x**3 + x**2 - x + 1)
print(sp.limit(f,x,-1))
```

```python
import numpy as np
import matplotlib.pyplot as plt

#    
x = np.linspace(-3, 3, 1000)
y = (x**3 + x**2 - x - 1) / (x**3 - x**2 - x + 1)

#    (x = 1  x = -1)
mask = (np.abs(x - 1) > 0.01) & (np.abs(x + 1) > 0.01)
x_clean = x[mask]
y_clean = y[mask]

#  
plt.figure(figsize=(10, 6))
plt.plot(x_clean, y_clean, 'b-', linewidth=2, label=r'$f(x) = \frac{x^3 + x^2 - x - 1}{x^3 - x^2 - x + 1}$')

#   
plt.axvline(x=1, color='red', linestyle='--', alpha=0.7, label='  x=1')
plt.axvline(x=-1, color='orange', linestyle='--', alpha=0.7, label='  x=-1')

#    x  -1
plt.plot(-1, 0, 'go', markersize=8, label='  x-1: 0')

#   
plt.axhline(y=1, color='purple', linestyle='--', alpha=0.7, label='  y=1')

#  
plt.xlim(-2, 2)
plt.ylim(-5, 5)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('    ')
plt.grid(True, alpha=0.3)
plt.legend()

#  
plt.tight_layout()
plt.show()

```

```python
from matplotlib.pyplot import*
import numpy as np
import sympy as sp
x = sp.Symbol('x')
f = (x**2)/(sp.cbrt(1+3*x) -1 -x)
print(sp.limit(f,x,0))
```

```python
from matplotlib.pyplot import*
from numpy import *

def f(x):
    return x**2 / (np.cbrt(1 + 3*x) - 1 - x)


x = np.linspace(-1, 1, 2000)

mask = np.abs(x) > 0.001
x_clean = x[mask]
y_clean = f(x_clean)
xlim(-0.5,0.5)
ylim(-3,1)

axhline(-1,color='red',linestyle='--')
axvline(0,color='green',linestyle='--')
grid()

plot(x_clean, y_clean, 'b-', linewidth=2, 
         label=r'$f(x) = \frac{x^2}{\sqrt[3]{1+3x} - 1 - x}$')


plot(0, -1, 'ro', markersize=8, markeredgewidth=2, 
         markerfacecolor='red', label='  x0: -1')
```

```python
#3.146

import numpy as np
import matplotlib.pyplot as plt
import math


def f(x):
    return ((x**(1/3) + 2*x**(1/6) + 1) / (x**(1/3) + 3)) ** (x**(1/6))


x = np.linspace(1, 10000, 1000)
y = f(x)
xlim(0,10000)
ylim(0,10)
limit = math.e**2
axhline(limit,color='red',linestyle='--')

plt.plot(x, y, 'b-', linewidth=2)

show()
```

```python
#4.56

import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return np.exp(-1 / (x**2 - 1))

x1 = np.linspace(-3, -1.0001, 1000)
x2 = np.linspace(-0.9999, 0.9999, 1000)
x3 = np.linspace(1.0001, 3, 1000)

y1 = f(x1)
y2 = f(x2)
y3 = f(x3)

plt.figure(figsize=(10, 6))
plt.plot(x1, y1, 'b', label='$f(x)$')
plt.plot(x2, y2, 'b')
plt.plot(x3, y3, 'b')

plt.axvline(x=-1, color='red', linestyle='--', alpha=0.7, label='x = 1')
plt.axvline(x=1, color='red', linestyle='--', alpha=0.7)

plt.ylim(-0.5, 5)
plt.xlim(-3, 3)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title(r' $f(x) = e^{-\frac{1}{x^2 - 1}}$')
plt.grid(True, linestyle=':')
plt.legend()
plt.show()
```
