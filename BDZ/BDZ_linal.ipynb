```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D

# Вершины треугольника
A = np.array([3, -4, 7])
B = np.array([-1, 2, -3])
C = np.array([2, -1, 0])

# Параметрические уравнения прямой AH (высота)
def line_ah(t):
    x = 3 + 1*t
    y = -4 + 2*t
    z = 7 + 1*t
    return x, y, z

# Параметрические уравнения прямой CM (медиана)
def line_cm(t):
    x = 2 + 1*t
    y = -1 - 1*t
    z = 0 + 5*t
    return x, y, z

# Построение
fig = plt.figure(figsize=(12, 10))
ax = fig.add_subplot(111, projection='3d')

# Треугольник ABC
vertices = np.array([A, B, C, A])
ax.plot(vertices[:, 0], vertices[:, 1], vertices[:, 2], 'bo-', 
        linewidth=3, markersize=8, label='Треугольник ABC')

# Высота AH
t_ah = np.linspace(-5, 5, 100)
x_ah, y_ah, z_ah = line_ah(t_ah)
ax.plot(x_ah, y_ah, z_ah, 'r--', linewidth=2, label='Высота AH')

# Медиана CM
t_cm = np.linspace(-3, 3, 100)
x_cm, y_cm, z_cm = line_cm(t_cm)
ax.plot(x_cm, y_cm, z_cm, 'g--', linewidth=2, label='Медиана CM')

# Подписи вершин
ax.text(A[0], A[1], A[2], '  A(3,-4,7)', fontsize=10)
ax.text(B[0], B[1], B[2], '  B(-1,2,-3)', fontsize=10)
ax.text(C[0], C[1], C[2], '  C(2,-1,0)', fontsize=10)

# Подписи прямых
ax.text(8, 6, 12, 'AH: (x-3)/1 = (y+4)/2 = (z-7)/1', color='red', fontsize=10)
ax.text(5, -6, 15, 'CM: (x-2)/1 = (y+1)/-1 = z/5', color='green', fontsize=10)

# Настройки
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('Треугольник ABC с высотой AH и медианой CM', fontsize=14)
ax.legend()

# Установим одинаковый масштаб по осям
max_range = max([A.ptp(), B.ptp(), C.ptp()]) * 1.2
mid_x = (A[0] + B[0] + C[0]) / 3
mid_y = (A[1] + B[1] + C[1]) / 3
mid_z = (A[2] + B[2] + C[2]) / 3

ax.set_xlim(mid_x - max_range/2, mid_x + max_range/2)
ax.set_ylim(mid_y - max_range/2, mid_y + max_range/2)
ax.set_zlim(mid_z - max_range/2, mid_z + max_range/2)

plt.tight_layout()
plt.show()
```
