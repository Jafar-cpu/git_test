```python
#3.25

import numpy as np
import matplotlib.pyplot as plt

# Уравнение плоскости: 21x + 20y + 6z - 43 = 0
# Выразим z: z = (43 - 21x - 20y) / 6

def plane(x, y):
    return (43 - 21*x - 20*y) / 6

# Параметризация прямой L
t = np.linspace(-3, 3, 100)
x_line = 1 + 2*t
y_line = -1 - 3*t
z_line = 7 + 3*t

# Создаём сетку для плоскости
x = np.linspace(-5, 5, 20)
y = np.linspace(-5, 5, 20)
X, Y = np.meshgrid(x, y)
Z = plane(X, Y)

# Точка M
M = np.array([1, 2, -3])

# Построение
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Плоскость
ax.plot_surface(X, Y, Z, alpha=0.5, color='cyan', label='Плоскость')

# Прямая L
ax.plot(x_line, y_line, z_line, color='red', linewidth=2, label='Прямая L')

# Точка M
ax.scatter(*M, color='blue', s=100, label='Точка M(1,2,-3)')

# Настройка
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.legend()
ax.set_title('Плоскость 21x + 20y + 6z - 43 = 0\nчерез точку M и прямую L')
plt.show()
```

```python
#3.55

import numpy as np
import matplotlib.pyplot as plt

# Вершины треугольника
A = np.array([3, -4, 7])
B = np.array([-1, 2, -3])
C = np.array([2, -1, 0])

# 1. Находим основание высоты H на прямой BC
BC_dir = C - B  # направляющий вектор BC
AB = A - B
BC_dir_dot = np.dot(BC_dir, BC_dir)
t_H = np.dot(AB, BC_dir) / BC_dir_dot
H = B + t_H * BC_dir

# 2. Находим середину стороны AB для медианы
M = (A + B) / 2  # середина AB

# 3. Параметрические уравнения для построения
# Сторона BC
t_vals = np.linspace(0, 1, 100)
BC_x = B[0] + BC_dir[0] * t_vals
BC_y = B[1] + BC_dir[1] * t_vals
BC_z = B[2] + BC_dir[2] * t_vals

# Медиана CM
u_vals = np.linspace(0, 1, 100)
median_x = C[0] + (M[0] - C[0]) * u_vals
median_y = C[1] + (M[1] - C[1]) * u_vals
median_z = C[2] + (M[2] - C[2]) * u_vals

# Высота AH (отрезок от A до H)
AH_x = [A[0], H[0]]
AH_y = [A[1], H[1]]
AH_z = [A[2], H[2]]

# Построение
fig = plt.figure(figsize=(18, 14))
ax = fig.add_subplot(111, projection='3d')

# Треугольник ABC
vertices = np.array([A, B, C, A])
ax.plot(vertices[:, 0], vertices[:, 1], vertices[:, 2], 'o-', 
        color='black', linewidth=3, markersize=10, label='Треугольник ABC')

# Сторона BC
ax.plot(BC_x, BC_y, BC_z, color='green', linewidth=2, 
        label='Сторона BC', alpha=0.7)

# Высота AH
ax.plot(AH_x, AH_y, AH_z, color='red', linewidth=5, 
        label='Высота AH ⟂ BC', alpha=0.9)

# Медиана CM
ax.plot(median_x, median_y, median_z, color='blue', linewidth=5, 
        label='Медиана CM', alpha=0.9, linestyle='-')

# Вершины и особые точки
ax.scatter(*A, color='darkblue', s=200, label='A(3,-4,7)')
ax.scatter(*B, color='darkblue', s=200, label='B(-1,2,-3)')
ax.scatter(*C, color='darkblue', s=200, label='C(2,-1,0)')
ax.scatter(*H, color='orange', s=250, marker='*', 
           label=f'H({H[0]:.1f},{H[1]:.1f},{H[2]:.1f}) - основание высоты')
ax.scatter(*M, color='purple', s=200, marker='s', 
           label=f'M({M[0]:.1f},{M[1]:.1f},{M[2]:.1f}) - середина AB')

# Подписи точек
ax.text(A[0], A[1], A[2], '  A', fontsize=14, fontweight='bold', color='darkblue')
ax.text(B[0], B[1], B[2], '  B', fontsize=14, fontweight='bold', color='darkblue')
ax.text(C[0], C[1], C[2], '  C', fontsize=14, fontweight='bold', color='darkblue')
ax.text(H[0], H[1], H[2], '  H\n(основание\nвысоты)', fontsize=11, fontweight='bold', color='orange')
ax.text(M[0], M[1], M[2], '  M\n(середина AB)', fontsize=11, fontweight='bold', color='purple')

# Показываем прямой угол у точки H
angle_size = 0.3
vec1 = (A - H) / np.linalg.norm(A - H) * angle_size
vec2 = (B - H) / np.linalg.norm(B - H) * angle_size
angle_points = np.array([H, H + vec1, H + vec1 + vec2, H + vec2, H])
ax.plot(angle_points[:, 0], angle_points[:, 1], angle_points[:, 2], 
        'g--', linewidth=2, alpha=0.7, label='Прямой угол')

# Настройка осей и внешнего вида
ax.set_xlabel('Ось X', fontsize=14, fontweight='bold')
ax.set_ylabel('Ось Y', fontsize=14, fontweight='bold')
ax.set_zlabel('Ось Z', fontsize=14, fontweight='bold')

ax.tick_params(axis='both', which='major', labelsize=12)
ax.legend(fontsize=13, loc='upper left', bbox_to_anchor=(0, 1))
ax.grid(True, alpha=0.3)

ax.set_title('Треугольник ABC: высота AH и медиана CM\n' + 
             'A(3,-4,7), B(-1,2,-3), C(2,-1,0)', 
             fontsize=18, fontweight='bold', pad=25)

# Устанавливаем удобный угол обзора
ax.view_init(elev=20, azim=35)
ax.set_box_aspect([1, 1, 1])

# Автоматическое определение границ
all_points = np.array([A, B, C, H, M])
margin = 1.5
ax.set_xlim([all_points[:, 0].min()-margin, all_points[:, 0].max()+margin])
ax.set_ylim([all_points[:, 1].min()-margin, all_points[:, 1].max()+margin])
ax.set_zlim([all_points[:, 2].min()-margin, all_points[:, 2].max()+margin])

plt.tight_layout()
plt.show()

# Дополнительная информация

print("\nУРАВНЕНИЯ:")
print("Высота AH: (x - 3)/1 = (y + 4)/2 = (z - 7)/1")
print("Медиана CM: (x - 2)/1 = (y + 1)/-1 = z/5")
```
