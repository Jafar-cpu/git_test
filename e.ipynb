```python
from matplotlib.pyplot import*
from numpy import*
from math import*

oa = array([-2,0])
ob = array([1,2])
oc = array([1,-1])

ab = ob - oa
bc = oc - ob
ac = oc - oa

print('AB:', ab)
print('BC:', bc)
print('AC:', ac)

grid()
ylim(-5,5)
xlim(-5,5)
arrow(-2,0,ab[0],ab[1],width=0.09, length_includes_head=True,facecolor='Yellow')
arrow(1,2,bc[0],bc[1],width=0.09, length_includes_head=True,facecolor='Red')
arrow(-2,0,ac[0],ac[1],width=0.09, length_includes_head=True,facecolor='Blue')
show()


```

```python
# Упражнение 2

from matplotlib.pyplot import*
from numpy import*
from math import*

a = array([-2,0])
b = array([3,4])
c = array([4,-1])
cos_ab = dot(a,b) / linalg.norm(a) / linalg.norm(b)
cos_ac = dot(a,c) / linalg.norm(a) / linalg.norm(c)
cos_bc = dot(b,c) / linalg.norm(b) / linalg.norm(c)

ab = b - a
bc = c - b
ac = c -a
grid()

arrow(-2,0,ab[0],ab[1],width=0.07, length_includes_head=True,facecolor='Yellow')

arrow(3,4,bc[0],bc[1],width=0.07, length_includes_head=True,facecolor='Yellow')
arrow(-2,0,ac[0],ac[1],width=0.07, length_includes_head=True,facecolor='Yellow')


show()
print('AB:', arccos(cos_ab))

print('AC:', arccos(cos_ac))
print('BC:', arccos(cos_bc))


```

```python
from matplotlib.pyplot import*
from numpy import*
ax = axes(projection='3d')
a = array([1,-2,0])
b = array([0,1,1])
c = array([1,2,2])
mt = array([[1,-2,0],
      [0,1,1],
      [1,2,2]])
op = linalg.det(mt)

if op !=  0:
    print(f'Определитель матрицы = {op}.\nТак как определитель != 0, то вектора линейна независимы, а значит образуют базис')

max_val = 7

ax.plot([0, max_val], [0, 0], [0, 0], color='Red')
ax.plot([0, 0], [0, max_val], [0, 0], color='Yellow')
ax.plot([0, 0], [0, 0], [0, max_val], color='Blue')

# Подписи осей
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

ax.set_xlim(0,7)

ax.set_zlim(0,7)
ax.set_ylim(-4,7)

ax.quiver(0,0,0,1,-2,0,length=1, normalize=False,color='Brown')
ax.quiver(0,0,0,0,1,1,length=1, normalize=False,color='Green')
ax.quiver(0,0,0,1,2,2,length=1, normalize=False,color='Black')


ax.axis()
grid()
show()













```

```python
from matplotlib.pyplot import*
from math import*
from numpy import*


ax =  axes(projection='3d')

a = array([1,3,-1])
b = array([2,-1,4])
c = array([5,0,3])
ab = b - a
ac = c - a
bc = c - b
z = cross(ab,ac)
ax.set_ylim([-4,4])
ax.set_xlim([-4,4])

ax.set_zlim([-4,4])
ax.quiver(*a,ab[0],ab[1],ab[2],color='Blue')
ax.quiver(*b,bc[0],bc[1],bc[2],color='Red')
ax.quiver(*a,ac[0],ac[1],ac[2],color='Green')

ax.quiver(*a,*z,color='Black')

abs_z = sqrt(z[0]**2 + z[1]**2 + z[2]**2)
print(abs_z)

ax.axis()
grid()
show()

```
