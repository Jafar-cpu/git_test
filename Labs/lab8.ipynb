```python
import sympy as sp


print("Упражнение 2: Символьное вычисление производных")
print("=" * 50)


x = sp.Symbol('x')


x0 = 0.5

print(f"Точка вычисления: x₀ = {x0}")
print()

# а) f(x) = arctg²(√x)
print("а) f(x) = arctg²(√x)")
print("-" * 30)


f1 = sp.atan(sp.sqrt(x))**2
print(f"f(x) = {f1}")

# Вычисляем производную
f1_deriv = sp.diff(f1, x)
print(f"f'(x) = {f1_deriv}")


f1_deriv_simplified = sp.simplify(f1_deriv)
print(f"f'(x) упрощенная = {f1_deriv_simplified}")

# значение производной в точке x₀ = 0.5
f1_deriv_value = f1_deriv_simplified.subs(x, x0)
print(f"f'({x0}) = {f1_deriv_value}")
#print(f"f'({x0}) численно ≈ {float(f1_deriv_value):.6f}")

print()

# б) f(x) = 3^(arcsin(x²))
print("б) f(x) = 3^(arcsin(x²))")
print("-" * 30)


f2 = 3**(sp.asin(x**2))
print(f"f(x) = {f2}")


f2_deriv = sp.diff(f2, x)
print(f"f'(x) = {f2_deriv}")

# Упрощаем производную
f2_deriv_simplified = sp.simplify(f2_deriv)
print(f"f'(x) упрощенная = {f2_deriv_simplified}")

# Вычисляем значение производной в точке x₀ = 0.5
f2_deriv_value = f2_deriv_simplified.subs(x, x0)
print(f"f'({x0}) = {f2_deriv_value}")
#print(f"f'({x0}) численно ≈ {float(f2_deriv_value):.6f}")

print()

print("Дополнительная информация:")

# Вычисляем также значения самих функций в точке x₀
print(f"\nЗначения функций в точке x₀ = {x0}:")
f1_value = f1.subs(x, x0)
f2_value = f2.subs(x, x0)
print(f"f₁({x0}) = arctg²(√{x0}) = {float(f1_value):.6f}")
print(f"f₂({x0}) = 3^(arcsin({x0}²)) = {float(f2_value):.6f}")

```

```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt

def plot_tangent(func_str, x0):

    x = sp.Symbol('x')
    f_expr = sp.sympify(func_str)
    
    f_x0 = f_expr.subs(x, x0)
    
    f_deriv = sp.diff(f_expr, x)
    
    f_deriv_x0 = f_deriv.subs(x, x0)
    
    tangent_expr = f_x0 + f_deriv_x0 * (x - x0)
    
    f_func = sp.lambdify(x, f_expr, 'numpy')
    tangent_func = sp.lambdify(x, tangent_expr, 'numpy')
    
    x_vals = np.linspace(x0 - 1, x0 + 1, 400)
    
    y_vals = f_func(x_vals)
    y_tangent = tangent_func(x_vals)
    
    plt.figure(figsize=(12, 8))
    
 
    plt.plot(x_vals, y_vals, 'b-', linewidth=2, label=f'$y = {sp.latex(f_expr)}$')
    

    plt.plot(x_vals, y_tangent, 'r--', linewidth=2, 
             label=f'Касательная в $x_0 = {x0:.3f}$')
    
   
    plt.plot(x0, f_x0, 'ro', markersize=8, label=f'Точка касания ({x0:.3f}, {float(f_x0):.3f})')
    

    plt.xlim(x0 - 1, x0 + 1)
    plt.xlabel('x', fontsize=12)
    plt.ylabel('y', fontsize=12)
    plt.title(f'Функция и касательная в точке $x_0 = {x0:.3f}$', fontsize=14)
    plt.grid(True, alpha=0.3)
    plt.legend(fontsize=10)
    plt.axhline(y=0, color='k', linewidth=0.5)
    plt.axvline(x=0, color='k', linewidth=0.5)
    
    
    print("=" * 60)
    print(f"Функция: y = {func_str}")
    print(f"Точка касания: x₀ = {x0}")
    print(f"Координаты точки касания: ({x0:.4f}, {float(f_x0):.4f})")
    print(f"Производная: f'(x) = {f_deriv}")
    print(f"Значение производной в точке x₀: f'({x0:.4f}) = {float(f_deriv_x0):.4f}")
    print(f"Уравнение касательной: y = {float(f_x0):.4f} + {float(f_deriv_x0):.4f}(x - {x0:.4f})")
    print(f"Упрощенное уравнение: y = {float(f_deriv_x0):.4f}x + {float(f_x0 - f_deriv_x0 * x0):.4f}")
    
    plt.tight_layout()
    plt.show()


print("Упражнение 3: Построение касательной к графику функции")
print("Функция: y = cos³(3x)")
print("Точка: x₀ = π/4")
print()


plot_tangent("cos(3*x)**3", np.pi/4)


```

```python
import numpy as np
import matplotlib.pyplot as plt

# Упражнение 4: Анализ движения тела по табличным данным
print("Упражнение 4: Анализ движения тела по табличным данным")
print("=" * 60)

# Данные из таблицы
t = np.array([0.7, 0.8, 0.9, 1.0, 1.1, 1.2, 1.3, 1.4, 1.5, 1.6, 1.7, 1.8, 1.9, 2.0])
s = np.array([2.3733, 2.5889, 2.7974, 3.0000, 3.1976, 3.3909, 3.5804, 
              3.7664, 3.9495, 4.1298, 4.3077, 4.4833, 4.6568, 4.8284])

print("Исходные данные:")
print("Время t:", t)
print("Путь s:", s)
print()

# а) Вычисление средних скоростей
print("а) Средние скорости на промежутках времени:")
print("-" * 50)

# Создаем матрицу для хранения результатов
delta_t_values = []
average_velocities = []

# Вычисляем средние скорости для разных промежутков
for i in range(len(t) - 1):
    delta_t = t[i+1] - t[i]
    delta_s = s[i+1] - s[i]
    v_avg = delta_s / delta_t
    
    delta_t_values.append(delta_t)
    average_velocities.append(v_avg)
    
    print(f"Δt = {delta_t:.1f}: v_ср = {v_avg:.4f}")

# Создаем матрицу результатов
velocity_matrix = np.array([delta_t_values, average_velocities])
print(f"\nМатрица средних скоростей:")
print("Δt:    ", velocity_matrix[0])
print("v_ср:  ", velocity_matrix[1])

# б) Построение графиков
print("\nб) Построение графиков:")
print("-" * 50)

# Создаем фигуру с двумя графиками
plt.figure(figsize=(14, 10))

# График 1: Зависимость пути от времени
plt.subplot(2, 1, 1)
plt.plot(t, s, 'bo-', linewidth=2, markersize=6, label='s(t) - пройденный путь')
plt.xlabel('Время t', fontsize=12)
plt.ylabel('Путь s', fontsize=12)
plt.title('Зависимость пройденного пути от времени', fontsize=14)
plt.grid(True, alpha=0.3)
plt.legend(fontsize=10)

# Добавляем точки на график
for i in range(len(t)):
    plt.annotate(f'{s[i]:.1f}', (t[i], s[i]), 
                textcoords="offset points", 
                xytext=(0,10), 
                ha='center', 
                fontsize=8)

# График 2: Мгновенная скорость (приближенно как средние скорости)
plt.subplot(2, 1, 2)

# Для мгновенной скорости используем средние точки интервалов
t_mid = (t[:-1] + t[1:]) / 2
v_instant = average_velocities

plt.plot(t_mid, v_instant, 'ro-', linewidth=2, markersize=6, label='v(t) - мгновенная скорость')
plt.xlabel('Время t', fontsize=12)
plt.ylabel('Скорость v', fontsize=12)
plt.title('Мгновенная скорость тела', fontsize=14)
plt.grid(True, alpha=0.3)
plt.legend(fontsize=10)

# Добавляем точки на график скорости
for i in range(len(t_mid)):
    plt.annotate(f'{v_instant[i]:.2f}', (t_mid[i], v_instant[i]), 
                textcoords="offset points", 
                xytext=(0,10), 
                ha='center', 
                fontsize=8)

plt.tight_layout()
plt.show()

# Дополнительный анализ: аппроксимация зависимости s(t)
print("\nДополнительный анализ:")
print("-" * 50)

# Попробуем аппроксимировать данные полиномом для получения аналитической зависимости
degree = 2  # степень полинома
coefficients = np.polyfit(t, s, degree)
poly_func = np.poly1d(coefficients)

print(f"Аппроксимирующая функция: s(t) = {coefficients[0]:.4f}t² + {coefficients[1]:.4f}t + {coefficients[2]:.4f}")

# Вычисляем производную (скорость) аналитически
v_analytical = np.polyder(poly_func)
print(f"Аналитическая скорость: v(t) = {v_analytical.coeffs[0]:.4f}t + {v_analytical.coeffs[1]:.4f}")

# Сравниваем численные и аналитические значения скорости
print("\nСравнение скоростей:")
print("Время\tЧисл. v\tАналит. v\tРазница")
for i in range(len(t_mid)):
    v_num = average_velocities[i]
    v_anal = v_analytical(t_mid[i])
    diff = abs(v_num - v_anal)
    print(f"{t_mid[i]:.1f}\t{v_num:.3f}\t{v_anal:.3f}\t\t{diff:.4f}")






# Вывод результатов в виде итоговой таблицы
print("\n" + "=" * 60)
print("ИТОГОВАЯ ТАБЛИЦА РЕЗУЛЬТАТОВ")
print("=" * 60)

print("\nИнтервалы времени и средние скорости:")
print("t нач.\tt кон.\tΔt\tΔs\tv_ср")
for i in range(len(t) - 1):
    print(f"{t[i]:.1f}\t{t[i+1]:.1f}\t{t[i+1]-t[i]:.1f}\t{s[i+1]-s[i]:.4f}\t{average_velocities[i]:.4f}")

print(f"\nСредняя скорость на всем промежутке: {np.mean(average_velocities):.4f}")
print(f"Максимальная скорость: {np.max(average_velocities):.4f}")
print(f"Минимальная скорость: {np.min(average_velocities):.4f}")

# Анализ характера движения
if abs(a_constant) < 0.01:
    print("\nХарактер движения: равномерное (ускорение ≈ 0)")
else:
    print(f"\nХарактер движения: равноускоренное (a = {a_constant:.4f})")
```

```python
# 6 задание 
import numpy as np
import matplotlib.pyplot as plt
from sympy import symbols, diff, lambdify

# Определяем символьную переменную и функцию Q(T)
T = symbols('T')
Q = 0.3965*T + 2.081e-3*T**2 - 5.024e-7*T**3

print("Функция Q(T):")
print(f"Q(T) = {Q}")

# Находим теплоемкость как производную dQ/dT
c = diff(Q, T)
print("\nФункция теплоемкости c(T):")
print(f"c(T) = {c}")

# Преобразуем символьные выражения в числовые функции
Q_func = lambdify(T, Q, 'numpy')
c_func = lambdify(T, c, 'numpy')

# Создаем массив значений температуры от 0 до 800°C
T_values = np.linspace(0, 800, 1000)

# Вычисляем значения Q и c
Q_values = Q_func(T_values)
c_values = c_func(T_values)

# Создаем графическое окно с двумя подграфиками
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10, 8))

# Первый график: Q(T)
ax1.plot(T_values, Q_values, 'b-', linewidth=2)
ax1.set_xlabel('Температура T, °C')
ax1.set_ylabel('Теплота Q, Дж')
ax1.set_title('Зависимость теплоты Q от температуры T для алмаза')
ax1.grid(True, alpha=0.3)
ax1.set_xlim(0, 800)

# Второй график: c(T)
ax2.plot(T_values, c_values, 'r-', linewidth=2)
ax2.set_xlabel('Температура T, °C')
ax2.set_ylabel('Теплоемкость c(T), Дж/(кг·°C)')
ax2.set_title('Зависимость теплоемкости c от температуры T для алмаза')
ax2.grid(True, alpha=0.3)
ax2.set_xlim(0, 800)

# Настраиваем расположение графиков
plt.tight_layout()
plt.show()

# Выведем несколько значений для проверки
print("\nПримеры значений:")
print("T (°C)\tQ (Дж)\tc(T) (Дж/(кг·°C))")
for temp in [0, 200, 400, 600, 800]:
    q_val = Q_func(temp)
    c_val = c_func(temp)
    print(f"{temp}\t{q_val:.2f}\t{c_val:.4f}")
```
