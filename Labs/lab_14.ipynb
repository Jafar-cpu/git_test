```python
#1

from numpy import*
from matplotlib.pyplot import*
from mpl_toolkits import mplot3d


x = linspace(-1,1,200)
y = linspace(-1,1,200)
[x,y] = meshgrid(x,y)

# 
z = lambda x,y: x**2 - y**2
Z = z(x,y)
fig = figure(figsize=(7,7))
ax = axes(projection='3d')
my_cmap = get_cmap('magma')

# 

surf = ax.plot_surface(x,y,Z,alpha = 0.8,cmap = my_cmap)
cset = ax.contourf(x,y,Z,zdir ='z',offset = np.min(Z),cmap=my_cmap)
cset = ax.contourf(x,y,Z,zdir ='x',offset =-2,cmap = my_cmap)
cset = ax.contourf(x,y,Z,zdir ='y',offset = 2,cmap = my_cmap)
fig.colorbar(surf,ax = ax,shrink = 0.5,aspect = 5)
ax.set_xlabel(' X')
ax.set_xlim(-2, 2)
ax.set_ylabel(' Y')
ax.set_ylim(-2, 2)
ax.set_zlabel(' Z')
ax.set_zlim(np.min(Z), np.max(Z))
ax.set_title('   $x^2 - y^2=z$  '); plt.show()
```

```python
# 2
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
from sympy import *
r = 1
fig = plt.figure(figsize =(9, 9))
ax = plt.axes(projection ='3d')

#   
phi = np.linspace(0, 2*np.pi, 50) 
tetta = np.linspace(-np.pi/2, np.pi/2, 50) 

phi,tetta= np.meshgrid(phi,tetta)

# 
x = r * np.cos(phi)*np.cos(tetta)
y = r * np.sin(phi)*np.cos(tetta)
z = r*np.sin(tetta)






my_cmap = plt.get_cmap('rainbow')
surf = ax.plot_surface(x,y,z,alpha = 0.8,cmap = my_cmap)
cset = ax.contourf(x,y,z,zdir ='z',offset = -2,cmap = my_cmap)
cset = ax.contourf(x,y,z,zdir ='x',offset =-4,cmap = my_cmap)
cset = ax.contourf(x,y,z,zdir ='y',offset = 2,cmap = my_cmap)
ax.set_xlabel(' X'); ax.set_xlim(-4, 4)
ax.set_ylabel(' Y'); ax.set_ylim(-2, 2)
ax.set_zlabel(' Z'); ax.set_zlim(-2, 2)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# 
R = 1.0  #  
phi_res = 50  #    
theta_res = 50  #    

#   
phi = np.linspace(0, 2*np.pi, phi_res)
theta = np.linspace(-np.pi/2, np.pi/2, theta_res)
phi_grid, theta_grid = np.meshgrid(phi, theta)

#   
x = R * np.cos(phi_grid) * np.cos(theta_grid)
y = R * np.sin(phi_grid) * np.cos(theta_grid)
z = R * np.sin(theta_grid)

#  3D 
fig = plt.figure(figsize=(12, 5))

#  :  
ax1 = fig.add_subplot(121, projection='3d')
surf = ax1.plot_surface(x, y, z, cmap='viridis', alpha=0.8, edgecolor='k', linewidth=0.5)
ax1.set_xlabel('X')
ax1.set_ylabel('Y')
ax1.set_zlabel('Z')
ax1.set_title(' : $x = R\cos\\varphi\cos\\theta$, $y = R\sin\\varphi\cos\\theta$, $z = R\sin\\theta$')

#   
fig.colorbar(surf, ax=ax1, shrink=0.5, aspect=10)

#    : x + y + z = R
x_flat = x.flatten()
y_flat = y.flatten()
z_flat = z.flatten()

#  x + y + z   
sphere_eq = x_flat**2 + y_flat**2 + z_flat**2

#    
max_error = np.max(np.abs(sphere_eq - R**2))
mean_error = np.mean(np.abs(sphere_eq - R**2))

print("    :")
print(f" : x + y + z = R,  R = {R}")
print(f" : R = {R**2}")
print(f"   R: {max_error:.2e}")
print(f"   R: {mean_error:.2e}")


```

```python

#3 
import numpy as np
import matplotlib.pyplot as plt

a, b, c = 2, 1, 1

phi = np.linspace(0, 2*np.pi, 50)
theta = np.linspace(-2, 2, 50)
phi, theta = np.meshgrid(phi, theta)

x = a * np.cos(phi) * np.cosh(theta)
y = b * np.sin(phi) * np.cosh(theta)
z = c * np.sinh(theta)

check = x**2/a**2 + y**2/b**2 - z**2/c**2
print(f" : {np.allclose(check, 1, atol=1e-10)}")

# 
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x, y, z, alpha=0.8, cmap='viridis')
ax.set_xlabel('X'), ax.set_ylabel('Y'), ax.set_zlabel('Z')
ax.set_title(' ', pad=20)

#  
print("\n :")
print("1.   XY:  x/a + y/b = 1 + z/c")
print("2.   XZ:  x/a - z/c = 1 - y/b")
print("3.   YZ:  y/b - z/c = 1 - x/a")

plt.show()
```

```python
#5
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

x = np.linspace(-3, 9, 100)
y = np.linspace(-3, 5, 50)
x_grid, y_grid = np.meshgrid(x, y)
z_grid = (x_grid**2 - 2*x_grid) / 2

t = np.linspace(-0.5, 2, 100)
line = Line(Point(8,-1,-3),Point(-4,3,3))
line = line.arbitrary_point()

x_line = 8 - 12*t
y_line = 4*t - 1
z_line = 6*t - 3

P1 = np.array([2, 1, 0])
P2 = np.array([-1, 2, 1.5])

fig = plt.figure(figsize=(12, 10))
ax = fig.add_subplot(111, projection='3d')

ax.plot_surface(x_grid, y_grid, z_grid, alpha=0.5, cmap='coolwarm')
ax.plot(x_line, y_line, z_line, 'b-', linewidth=3, label='  A  B')

ax.scatter([8, -4], [-1, 3], [-3, 3], color='red', s=100, label=' A, B')
ax.scatter([P1[0], P2[0]], [P1[1], P2[1]], [P1[2], P2[2]], 
           color='green', s=150, marker='*', label=' ')

ax.set_xlabel('X'), ax.set_ylabel('Y'), ax.set_zlabel('Z')
ax.set_title('   $x^2 - 2x = 2z$  ')
ax.legend()
ax.view_init(30, -60)

plt.tight_layout()
plt.show()

print(" :")
print(f"P({P1[0]}, {P1[1]}, {P1[2]})")
print(f"P({P2[0]}, {P2[1]}, {P2[2]})")
```
