```python
import numpy as np

#Метод обратной матрицы
A = np.array([[4, 3, 2], [-2, 2, 3],[3, -5, 2]])
B = np.array([25, -10, -4])
#Исследуем систему на совместность
D = np.column_stack((A,B)) 
# добавляем к матрице А столбец В
n=3
rkA=np.linalg.matrix_rank(A)
rkD=np.linalg.matrix_rank(D)
print('Проверка совместности:',[rkA,rkD,n])

X1=np.linalg.inv(A).dot(B)
print('Ответ:',X1)

#Метод Крамера
DA=np.linalg.det(A)
D1=A.copy();D1[:,0]=B
D11=np.linalg.det(D1)
D2=A.copy();D2[:,1]=B
D22=np.linalg.det(D2)
D3=A.copy();D3[:,2]=B
D33=np.linalg.det(D3)
X2=[D11/DA,D22/DA,D33/DA]
print('Ответ:',X2)
#Метод solve
X3 = np.linalg.solve(A,B)
print('Ответ:',X3)
```
