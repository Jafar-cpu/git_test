```python
# Упражнение 1

import sympy as sp
from sympy.abc import pi
import numpy as np
sumbl = [sp.cos((sp.pi*i)/2) for i in range(1,11)]
a = [np.cos((np.pi*i)/2) for i in range(1,11)]
print(sumbl)
print(a)
print(a==sumbl)
# sp.cos хранит точное значение
# np.cos хранит приблеженнное значение 
```

```python
# Упражнение 2
from sympy import*
from sympy.abc import x
v = (x**7 + 3*x**2 - 4)/(x-1)
v1 = (x**6  + x**5 + 3*x**4 + 5*x**3 - 6*x**2 + 4*x - 8)/(x**2 - 1)
pprint(v)
pprint(v1)
print('')
print('---- Разложение ----')
print(simplify(v))
print(simplify(v1))
print('')
print('---- Упрощение ----')
print(factor(v))
print(factor(v1))
```
