```python
# Упражнение 4
from matplotlib.pyplot import*
from numpy import*
z = 2*sqrt(3)-2j
r = abs(z)
phi = angle(z)
k = arange(1,7)
zroot = r**(1/8)*(cos((phi+2.0*pi*k)/8) + 1j*sin((phi+2.0*pi*k)/8))

angles = angle(zroot)
sort_idx = argsort(angles)
vertices = zroot[sort_idx]

vertices_closed = append(vertices, vertices[0])

print(zroot)
plot(real(zroot), imag(zroot),'*g',label='Корни уравнения')
axhline(y=0,color='k')
axvline(x=0,color='k')

axis('equal')

title('Корни уравнения и построенный многоугольник')
plot(vertices_closed.real, vertices_closed.imag, 'r-', linewidth=1, linestyle='--', label='Многоугольник')
t = linspace(0,2*pi,100)
plot(r**(1/8)*cos(t),r**(1/8)*sin(t),':')

legend()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Задаем уравнение, например: z^3 - 1 = 0 (корни — корни третьей степени из 1)
degree = 3
roots = np.array([np.exp(2j*np.pi*k/degree) for k in range(degree)])

# Координаты корней
x = roots.real
y = roots.imag

# Создаем фигуру и оси
fig, ax = plt.subplots()

# Рисуем точки корней
ax.plot(x, y, 'o', label='Корни у уравнения')

# Радиус окружности, на которой лежат точки
radius = 1

# Изобразить окружность
circle = plt.Circle((0,0), radius, color='gray', linestyle='--', fill=False)
ax.add_artist(circle)

# Построить правильный многоугольник (соединяем вершины)
# Для этого берем точки и соединяем их в порядке, образующем правильный многоугольник
# В данном случае корни уже расположены на окружности равномерно

# Сортируем точки по углу, чтобы соединять по порядку
angles = np.angle(roots)
sort_idx = np.argsort(angles)
vertices = roots[sort_idx]

# Замыкаем многоугольник, добавляя первую точку в конец
vertices_closed = np.append(vertices, vertices[0])

# Рисуем многоугольник
ax.plot(vertices_closed.real, vertices_closed.imag, 'r-', linewidth=1, linestyle='--', label='Многоугольник')

# Настраиваем оси
ax.set_aspect('equal', adjustable='box')
ax.grid(True, which='both', color='gray', linewidth=0.5)

# Подписываем оси
ax.set_xlabel('Re', fontsize=12)
ax.set_ylabel('Im', fontsize=12)
ax.set_title('Корни уравнения и построенный многоугольник')

# Устанавливаем одинаковый масштаб по осям
ax.set_xlim(-1.5, 1.5)
ax.set_ylim(-1.5, 1.5)

ax.legend()
plt.show()
```
