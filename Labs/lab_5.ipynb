```python
# Упражнение 4
from matplotlib.pyplot import*
from numpy import*
z = 2*sqrt(3)-2j
r = abs(z)
phi = angle(z)
k = arange(1,7)
zroot = r**(1/8)*(cos((phi+2.0*pi*k)/8) + 1j*sin((phi+2.0*pi*k)/8))

angles = angle(zroot)
sort_idx = argsort(angles)
vertices = zroot[sort_idx]

vertices_closed = append(vertices, vertices[0])

print(zroot)
plot(real(zroot), imag(zroot),'*g',label='Точки')
axhline(y=0,color='k')
axvline(x=0,color='k')

axis('equal')
title('Точки и построенный многоугольник')
plot(vertices_closed.real, vertices_closed.imag, 'r-', linewidth=1, linestyle='--', label='Многоугольник')
t = linspace(0,2*pi,100)
plot(r**(1/8)*cos(t),r**(1/8)*sin(t),':')

legend()
```

```python
import matplotlib.pyplot as plt
import numpy as np

def plot_arg_const(arg_value, r_min=0, r_max=2, num_points=300):
    r = np.linspace(r_min, r_max, num_points)
    z = r * np.exp(1j * arg_value)
    plt.plot(z.real, z.imag, '.', label=f'arg(z) = {arg_value:.2f}')

# Пример вызова для arg=Pi/4
plot_arg_const(5*np.pi/6)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.gca().set_aspect('equal')
plt.grid(True, linestyle='--')
plt.xlabel('Re')
plt.ylabel('Im')
plt.title('Множество точек для постоянного аргумента')
plt.legend()
plt.show()
```
