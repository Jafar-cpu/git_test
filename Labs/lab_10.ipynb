```python
# Упражнение 3
from sympy import Point, Line
import matplotlib.pyplot as plt
import numpy as np


L = Line(Point(0, 0), Point(2, 3))
print(f"Уравнение исходной прямой L: {L.equation()}")


L1 = L.parallel_line(Point(-2, 2))
print(f"Уравнение параллельной прямой L1: {L1.equation()}")

fig, ax = plt.subplots(figsize=(10, 8))
x = np.linspace(-5, 5, 100)
y_L = (3/2) * x
y_L1 = (3/2) * x + 5


ax.plot(x, y_L, 'b-', linewidth=2, label=f'L: {L.equation()}')
ax.plot(x, y_L1, 'r-', linewidth=2, label=f'L1: {L1.equation()}')


points = [(0, 0), (2, 3), (-2, 2)]
colors = ['blue', 'blue', 'red']
labels = ['(0,0)', '(2,3)', '(-2,2)']

for (x, y), color, label in zip(points, colors, labels):
    ax.plot(x, y, 'o', markersize=8, color=color, label=label)


ax.set_xlim(-5, 5)
ax.set_ylim(-5, 5)
ax.axhline(y=0, color='k', linewidth=0.5)
ax.axvline(x=0, color='k', linewidth=0.5)
ax.grid(True, alpha=0.3)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_title('Параллельные прямые')
ax.legend()
ax.set_aspect('equal')

plt.show()
```

```python
# Упражнение 4 и 6

from sympy import Point, Line
import matplotlib.pyplot as plt
import numpy as np


L = Line(Point(0, 0), Point(2, 3))
print(f"Уравнение исходной прямой L: {L.equation()}")


L2 = L.perpendicular_line(Point(-2, 2))
print(f"Уравнение перпендикулярной прямой L2: {L2.equation()}")


intersection = L.intersection(L2)
if intersection:
    intersect_point = intersection[0]
    print(f"Точка пересечения прямых: ({intersect_point.x}, {intersect_point.y})")


fig, ax = plt.subplots(figsize=(10, 8))
x = np.linspace(-5, 5, 100)
y_L = (3/2) * x
y_L2 = (-2/3) * x + (2/3)

ax.plot(x, y_L, 'b-', linewidth=2, label=f'L: y = (3/2)x')
ax.plot(x, y_L2, 'g-', linewidth=2, label=f'L2: y = (-2/3)x + 2/3')

points = [(0, 0), (2, 3), (-2, 2)]
colors = ['blue', 'blue', 'green']
labels = ['(0,0)', '(2,3)', '(-2,2)']

for (x, y), color, label in zip(points, colors, labels):
    ax.plot(x, y, 'o', markersize=8, color=color, label=label)


if intersection:
    ix, iy = float(intersect_point.x), float(intersect_point.y)
    ax.plot(ix, iy, 'ro', markersize=10, label=f'Пересечение: ({ix:.2f}, {iy:.2f})')
    
    ax.axhline(y=iy, color='red', linestyle='--', alpha=0.3)
    ax.axvline(x=ix, color='red', linestyle='--', alpha=0.3)


ax.set_xlim(-5, 5)
ax.set_ylim(-5, 5)
ax.axhline(y=0, color='k', linewidth=0.5)
ax.axvline(x=0, color='k', linewidth=0.5)
ax.grid(True, alpha=0.3)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_title('Перпендикулярные прямые и их точка пересечения')
ax.legend()
ax.set_aspect('equal')

plt.show()

```

```python
from sympy import Point, Line
import matplotlib.pyplot as plt
import numpy as np


M0 = Point(0.6, -0.4)  
q = (-3, 3)            
start_point = Point(0, -2)  
M1 = Point(M0.x + q[0], M0.y + q[1])
line = Line(M0, M1)

print(f"Общее уравнение прямой: {line.equation()}")
fig, ax = plt.subplots(figsize=(12, 10))
x_line = np.linspace(-5, 5, 100)
A, B, C = line.coefficients
if B != 0:
    y_line = (-A*x_line - C) / B
else:
   
    y_line = np.linspace(-5, 5, 100)
    x_line = np.full_like(y_line, -C/A)


ax.plot(x_line, y_line, 'b-', linewidth=2, label=f'Прямая: {line.equation()}')
ax.plot(M0.x, M0.y, 'ro', markersize=8, label=f'Точка M₀({M0.x}, {M0.y})')
ax.quiver(start_point.x, start_point.y, q[0], q[1], 
          angles='xy', scale_units='xy', scale=1, 
          color='red', width=0.008, label=f'Вектор q{q}')

ax.plot(start_point.x, start_point.y, 'go', markersize=6, label=f'Начало вектора ({start_point.x}, {start_point.y})')

ax.set_xlim(-5, 5)
ax.set_ylim(-5, 5)
ax.axhline(y=0, color='k', linewidth=0.5)
ax.axvline(x=0, color='k', linewidth=0.5)
ax.grid(True, alpha=0.3)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_title('Прямая через точку M₀, параллельная вектору q')
ax.legend()
ax.set_aspect('equal')
ax.grid(True, alpha=0.3)
plt.show()
```

```python
# Упражнение 8

import matplotlib.pyplot as plt
import numpy as np
from sympy import Point, Line


A = Point(2, -2)
B = Point(3, 5)
C = Point(6, 1)
AC = Line(A, C)
h_b = AC.perpendicular_line(B)
h_B = h_b.equation()
h_B_len = B.distance(AC)

print('Уравнение прямой на которой лежит высота из В:',h_B)
print('Длина этой высоты:',h_B_len)

print('*'*100)
# Медиана из А и ее длина
M_BC = Point((B.x + C.x)/2, (B.y + C.y)/2)  # середина BC
median_A = Line(A, M_BC)
med_A = median_A.equation()
median_length = A.distance(M_BC)

print('Уравнение прямой на которой лежит медиана из А:',med_A)
print('Длина этой медианы:',median_length)

print('*'*100)
cd = Line(C,D)
CD_e = cd.equation()
print('Уравнение прямой на которой лежит биссектриса из С:', CD_e)

fig, ax = plt.subplots(figsize=(12, 10))
triangle_x = [A.x, B.x, C.x, A.x]
triangle_y = [A.y, B.y, C.y, A.y]
ax.plot(triangle_x, triangle_y, 'k-', linewidth=2, label='Треугольник ABC')


ax.plot(A.x, A.y, 'ro', markersize=8, label='A(2,-2)')
ax.plot(B.x, B.y, 'ro', markersize=8, label='B(3,5)')
ax.plot(C.x, C.y, 'ro', markersize=8, label='C(6,1)')


ax.text(A.x - 0.3, A.y - 0.4, 'A', fontsize=12, fontweight='bold')
ax.text(B.x - 0.3, B.y + 0.3, 'B', fontsize=12, fontweight='bold')
ax.text(C.x + 0.2, C.y, 'C', fontsize=12, fontweight='bold')


AC = Line(A, C)
height_B = AC.perpendicular_line(B)
foot_B = AC.projection(B)
ax.plot([B.x, foot_B.x], [B.y, foot_B.y], 'b--', linewidth=2, label='Высота из B')
ax.plot(foot_B.x, foot_B.y, 'bs', markersize=6)


M_BC = Point((B.x + C.x)/2, (B.y + C.y)/2)  # середина BC
ax.plot([A.x, M_BC.x], [A.y, M_BC.y], 'g--', linewidth=2, label='Медиана из A')
ax.plot(M_BC.x, M_BC.y, 'gs', markersize=6)
ax.text(M_BC.x + 0.2, M_BC.y, 'M', fontsize=10)


CA = A.distance(C)
CB = B.distance(C)
x_D = (A.x * CB + B.x * CA) / (CA + CB)
y_D = (A.y * CB + B.y * CA) / (CA + CB)
D = Point(x_D, y_D)

ax.plot([C.x, D.x], [C.y, D.y], 'r--', linewidth=2, label='Биссектриса из C')
ax.plot(D.x, D.y, 'rs', markersize=8)  # Увеличили размер точки D
ax.text(D.x + 0.2, D.y, 'D', fontsize=12, fontweight='bold', color='red') 
ax.text(D.x - 0.8, D.y - 0.4, f'D({D.x:.1f}, {D.y:.1f})', 
        fontsize=10, color='red', bbox=dict(boxstyle="round,pad=0.3", facecolor="white", alpha=0.8))

ax.set_xlim(0, 7)
ax.set_ylim(-3, 6)
ax.axhline(y=0, color='k', linewidth=0.5, alpha=0.3)
ax.axvline(x=0, color='k', linewidth=0.5, alpha=0.3)
ax.grid(True, alpha=0.3)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_title('Треугольник с высотой, медианой и биссектрисой', fontsize=14)
ax.legend(loc='upper right')
ax.set_aspect('equal')

plt.tight_layout()
plt.show()
```
