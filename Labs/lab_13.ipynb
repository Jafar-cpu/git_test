```python
#4
import numpy as np
import matplotlib.pyplot as plt

a = np.sqrt(2)   
b = np.sqrt(8)   
h = 2
k = 0

t = np.linspace(-3, 3, 400)

x_upper = h + b * np.sinh(t)
y_upper = a * np.cosh(t)

x_lower = h + b * np.sinh(t)
y_lower = -a * np.cosh(t)

t_points = [-1, 0, 1]
x_points = h + b * np.sinh(t_points)
y_points_upper = a * np.cosh(t_points)
y_points_lower = -a * np.cosh(t_points)

plt.figure(figsize=(8, 8))
plt.plot(x_upper, y_upper, 'b-', label=' ')
plt.plot(x_lower, y_lower, 'r-', label=' ')

plt.scatter(x_points, y_points_upper, color='green', s=100, zorder=5)
for i, tp in enumerate(t_points):
    plt.annotate(f't={tp:.1f}', (x_points[i], y_points_upper[i]),
                 textcoords="offset points", xytext=(0, 10), ha='center')

plt.scatter(x_points, y_points_lower, color='orange', s=100, zorder=5)
for i, tp in enumerate(t_points):
    plt.annotate(f't={tp:.1f}', (x_points[i], y_points_lower[i]),
                 textcoords="offset points", xytext=(0, -15), ha='center')

plt.xlabel('x')
plt.ylabel('y')
plt.title(': $\\frac{y^2}{2} - \\frac{(x-2)^2}{8} = 1$')
plt.grid(True)
plt.legend()
plt.axis('equal')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

#  : 4x^2 + 9y^2 - 8x - 36y + 4 = 0
# : (x-1)^2/9 + (y-2)^2/4 = 1

center = (1, 2)
a = 3
b = 2
c = np.sqrt(a**2 - b**2)  # sqrt(5)

f1 = (center[0] + c, center[1])
f2 = (center[0] - c, center[1])

ecc = c / a
directrix_right = center[0] + a / ecc
directrix_left = center[0] - a / ecc

theta = np.linspace(0, 2 * np.pi, 300)
x_ellipse = center[0] + a * np.cos(theta)
y_ellipse = center[1] + b * np.sin(theta)

plt.figure(figsize=(8, 8))
plt.plot(x_ellipse, y_ellipse, 'b-', label='')
plt.plot(center[0], center[1], 'ko', label=' (1, 2)')
plt.plot(f1[0], f1[1], 'ro', label=' $F_1$')
plt.plot(f2[0], f2[1], 'ro', label=' $F_2$')
plt.axvline(directrix_right, color='gray', linestyle='--', label='')
plt.axvline(directrix_left, color='gray', linestyle='--')

plt.text(f1[0], f1[1] + 0.2, f'({f1[0]:.2f}, {f1[1]:.2f})', ha='center')
plt.text(f2[0], f2[1] + 0.2, f'({f2[0]:.2f}, {f2[1]:.2f})', ha='center')

plt.xlabel('x')
plt.ylabel('y')
plt.title(': $4x^2 + 9y^2 - 8x - 36y + 4 = 0$\n : $(x-1)^2/9 + (y-2)^2/4 = 1$')
plt.grid(True)
plt.legend()
plt.axis('equal')
plt.show()

print(":", center)
print("a =", a, ", b =", b)
print("c = (a - b) =", round(c, 4))
print("  = c/a =", round(ecc, 4))
print(":", (round(f1[0], 4), f1[1]), (round(f2[0], 4), f2[1]))
print(": x =", round(directrix_left, 4), " x =", round(directrix_right, 4))
```

```python
def f(A, B, D, E, F):
    """
      Ax^2 + By^2 + Dx + Ey + F = 0
      A(x - h)^2 + B(y - k)^2 = R
     (A, B, h, k, R)
    """
    h = 0
    k = 0
    R = -F  #  F    
    
    if A != 0:
        h = -D / (2 * A)
        # Dx = A*(x^2 + (D/A)x)
        #  : (D/(2A))^2 * A = D^2/(4A)
        R += A * (h**2)
    else:
        #   x  (   )
        pass
    
    if B != 0:
        k = -E / (2 * B)
        R += B * (k**2)
        
    return A, B, h, k, R
def f1(A, B, h, k, R):
    """
         .
    """
    parts = []
    if A != 0:
        parts.append(f"{A}(x - {h:.3f})^2")
    if B != 0:
        parts.append(f"{B}(y - {k:.3f})^2")
    left = " + ".join(parts)
    return f"{left} = {R:.3f}"
A, B, D, E, F = 4, 9, -8, -36, 4
A1, B1, h, k, R = f(A, B, D, E, F)
print("  . 5:")
print(f": {A}x^2 + {B}y^2 + {D}x + {E}y + {F} = 0")
print(f"  : {f1(A1, B1, h, k, R)}")
print(f"h = {h}, k = {k}, R = {R}")    
```
