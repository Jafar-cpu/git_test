```python
# 1 

import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import fsolve

#  
def f(x):
    return (x**4 + 2*x**2 + x + 2) / (x**3 + 2*x**2 + x + 2)

#   x   
x1 = np.linspace(-10, -2.1, 1000)  #   
x2 = np.linspace(-1.9, 5, 1000)    #   

#  y 
y1 = f(x1)
y2 = f(x2)

#  
plt.figure(figsize=(12, 8))

#   
plt.plot(x1, y1, 'b-', linewidth=2, label='f(x)')
plt.plot(x2, y2, 'b-', linewidth=2)

# 
x_asymptote = -2
plt.axvline(x=x_asymptote, color='red', linestyle='--', alpha=0.7, label='  x=-2')

#   y = x - 2
x_asymptote_line = np.linspace(-8, 5, 100)
y_asymptote_line = x_asymptote_line - 2
plt.plot(x_asymptote_line, y_asymptote_line, 'g--', alpha=0.7, label='  y=x-2')

#    
def derivative_eq(x):
    return x**6 + 4*x**5 + x**4 + 6*x**3 - 6*x**2

#   
x_extremum1 = fsolve(derivative_eq, -4.15)[0]
x_extremum2 = fsolve(derivative_eq, 0.68)[0]

y_extremum1 = f(x_extremum1)
y_extremum2 = f(x_extremum2)

plt.plot(x_extremum1, y_extremum1, 'ro', markersize=8, label=f' ({x_extremum1:.2f}, {y_extremum1:.2f})')
plt.plot(x_extremum2, y_extremum2, 'go', markersize=8, label=f' ({x_extremum2:.2f}, {y_extremum2:.2f})')

#  (0, f(0))
plt.plot(0, f(0), 'mo', markersize=8, label=f' (0, {f(0):.1f})')

#  
plt.xlim(-10, 5)
plt.ylim(-15, 10)
plt.grid(True, alpha=0.3)
plt.axhline(y=0, color='k', linewidth=0.5)
plt.axvline(x=0, color='k', linewidth=0.5)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('  f(x) = (x + 2x + x + 2) / (x + 2x + x + 2)')
plt.legend()

#       
plt.figure(figsize=(10, 6))
x_detail1 = np.linspace(-3, -2.01, 500)
x_detail2 = np.linspace(-1.99, 1, 500)
plt.plot(x_detail1, f(x_detail1), 'b-', linewidth=2)
plt.plot(x_detail2, f(x_detail2), 'b-', linewidth=2)
plt.axvline(x=-2, color='red', linestyle='--', alpha=0.7, label=' x=-2')
plt.plot(x_extremum2, y_extremum2, 'go', markersize=8, label=f' ({x_extremum2:.2f}, {y_extremum2:.2f})')
plt.plot(0, f(0), 'mo', markersize=8, label=f' (0, {f(0):.1f})')
plt.xlim(-3, 1)
plt.ylim(-5, 5)
plt.grid(True, alpha=0.3)
plt.axhline(y=0, color='k', linewidth=0.5)
plt.axvline(x=0, color='k', linewidth=0.5)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('    x = -2')
plt.legend()

plt.tight_layout()
plt.show()

#   
print("  :")
print(f" : x = -2")
print(f" : y = x - 2")
print(f" : ({x_extremum1:.4f}, {y_extremum1:.4f})")
print(f" : ({x_extremum2:.4f}, {y_extremum2:.4f})")
print(f"    Y: (0, {f(0):.4f})")
print(f":")
print(f"  lim(x-2) f(x) = -")
print(f"  lim(x-2) f(x) = +")
print(f"  lim(x-) f(x) = - (  y=x-2 )")
print(f"  lim(x+) f(x) = + (  y=x-2 )")
```

```python
# 3  

import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import fsolve
from scipy.misc import derivative

#  
def f(x):
    return np.sqrt((x**3 - 2*x**2) / (x - 3))

#  
plt.figure(figsize=(14, 10))

#  ,   
#    (x - 2x)/(x-3) >= 0
#   :
# x(x-2)/(x-3) >= 0

#  :
# x  [0, 2]  (3, +)
x_domain1 = np.linspace(0, 2, 1000)  # [0, 2]
x_domain2 = np.linspace(3.1, 8, 1000)  # (3, +)

#   
y1 = f(x_domain1)
y2 = f(x_domain2)

# 1.   
plt.plot(x_domain1, y1, 'b-', linewidth=2, label='f(x)', zorder=1)
plt.plot(x_domain2, y2, 'b-', linewidth=2, zorder=1)

# 2. 
#  
plt.axvline(x=3, color='red', linestyle='--', alpha=0.7, label='  x=3', zorder=2)

# 3.     
def find_zeros():
    zeros = []
    # f(x) = 0    = 0: (x - 2x)/(x-3) = 0
    # x - 2x = x(x - 2) = 0
    zero1 = 0.0
    zero2 = 2.0
    
    # ,     
    zeros = [zero1, zero2]
    
    for zero in zeros:
        plt.plot(zero, 0, 'ro', markersize=10, zorder=3, markeredgecolor='black')
    
    return zeros

zeros = find_zeros()

# 4.  
def find_extremums():
    #  
    def f_prime(x):
        return derivative(f, x, dx=1e-6)
    
    #      
    initial_guesses = [0.5, 1.0, 1.5, 4, 5, 6]
    extremums = []
    
    for guess in initial_guesses:
        try:
            critical_point = fsolve(f_prime, guess)[0]
            # ,        near 
            if ((0 <= critical_point <= 2) or critical_point > 3.1) and abs(critical_point - 3) > 0.2:
                #   
                is_unique = True
                for existing, _, _ in extremums:
                    if abs(critical_point - existing) < 0.1:
                        is_unique = False
                        break
                if is_unique:
                    #   
                    second_deriv = derivative(f_prime, critical_point, dx=1e-6)
                    if second_deriv > 0:
                        extremum_type = ''
                        color = 'green'
                        marker = '^'
                    else:
                        extremum_type = ''
                        color = 'red'
                        marker = 'v'
                    
                    y_val = f(critical_point)
                    extremums.append((critical_point, y_val, extremum_type))
                    plt.plot(critical_point, y_val, marker, color=color, 
                            markersize=12, markeredgecolor='black', markeredgewidth=1, zorder=3)
        except:
            continue
    
    return extremums

extremums = find_extremums()

# 5.   
def find_inflection_points():
    #  
    def f_double_prime(x):
        return derivative(f, x, dx=1e-6, n=2)
    
    #   
    initial_guesses = [0.7, 1.3, 4.5, 6]
    inflection_points = []
    
    for guess in initial_guesses:
        try:
            inflection_point = fsolve(f_double_prime, guess)[0]
            # ,     
            if ((0 <= inflection_point <= 2) or inflection_point > 3.1) and abs(inflection_point - 3) > 0.2:
                #   
                is_unique = True
                for existing, _ in inflection_points:
                    if abs(inflection_point - existing) < 0.1:
                        is_unique = False
                        break
                if is_unique:
                    y_val = f(inflection_point)
                    inflection_points.append((inflection_point, y_val))
                    plt.plot(inflection_point, y_val, 's', color='orange', 
                            markersize=10, markeredgecolor='black', markeredgewidth=1, zorder=3)
        except:
            continue
    
    return inflection_points

inflection_points = find_inflection_points()

# 6.   ()
x_undefined1 = np.linspace(-2, -0.1, 50)
x_undefined2 = np.linspace(2.1, 2.9, 50)
plt.axvspan(-2, 0, alpha=0.2, color='gray', label=' ')
plt.axvspan(2, 3, alpha=0.2, color='gray')

# 7.  
plt.xlim(-2, 8)
plt.ylim(-1, 8)
plt.grid(True, alpha=0.3, zorder=0)
plt.axhline(y=0, color='k', linewidth=0.5, zorder=0)
plt.axvline(x=0, color='k', linewidth=0.5, zorder=0)
plt.xlabel('x', fontsize=12)
plt.ylabel('f(x)', fontsize=12)
plt.title(' : $f(x) = \\sqrt{\\frac{x^3 - 2x^2}{x - 3}}$', fontsize=14)

# 8.    
legend_elements = [
    plt.Line2D([0], [0], color='blue', lw=2, label='f(x)'),
    plt.Line2D([0], [0], color='red', linestyle='--', label='  x=3'),
    plt.Rectangle((0,0), 1, 1, facecolor='gray', alpha=0.2, label=' '),
    plt.Line2D([0], [0], marker='o', color='red', markersize=8, label=' ', linestyle='None'),
    plt.Line2D([0], [0], marker='v', color='red', markersize=8, label='', linestyle='None'),
    plt.Line2D([0], [0], marker='^', color='green', markersize=8, label='', linestyle='None'),
    plt.Line2D([0], [0], marker='s', color='orange', markersize=8, label=' ', linestyle='None')
]

plt.legend(handles=legend_elements, loc='upper right', bbox_to_anchor=(1, 1))

# 9.     
info_text = "  :\n\n"

#  
info_text += " :\n"
info_text += "  x  [0, 2]  (3, +)\n"

# 
info_text += "\n :\n"
for i, zero in enumerate(zeros):
    info_text += f"   {i+1}: ({zero:.1f}, 0.0)\n"

# 
info_text += "\n:\n"
for i, (x, y, e_type) in enumerate(extremums):
    info_text += f"  {e_type}: ({x:.2f}, {y:.2f})\n"

#  
info_text += "\n :\n"
for i, (x, y) in enumerate(inflection_points):
    info_text += f"   {i+1}: ({x:.2f}, {y:.2f})\n"

# 
info_text += "\n:\n"
info_text += "  : x = 3\n"
info_text += "  : "

plt.text(0.02, 0.98, info_text, transform=plt.gca().transAxes, 
         verticalalignment='top', bbox=dict(boxstyle='round', facecolor='wheat', alpha=0.8),
         fontsize=10, family='monospace')

plt.tight_layout()
plt.show()

# 10.     
print("="*60)
print("  ")
print("="*60)
print(f": f(x) = ((x - 2x)/(x - 3))")
print(f" : x  [0, 2]  (3, +)")
print(f" : y  0")

print(f"\n :")
for i, zero in enumerate(zeros):
    print(f"  x_{i+1} = {zero:.1f}")

print(f"\n:")
for i, (x, y, e_type) in enumerate(extremums):
    print(f"  {e_type}: x = {x:.4f}, y = {y:.4f}")

print(f"\n :")
for i, (x, y) in enumerate(inflection_points):
    print(f"   {i+1}: x = {x:.4f}, y = {y:.4f}")

print(f"\n:")
print(f"  : x = 3")

print(f"\n  :")
print(f"  lim(x0) f(x) = {f(0.001):.2f}")
print(f"  lim(x2) f(x) = {f(1.999):.2f}")
print(f"  lim(x3) f(x) = +")
print(f"  lim(x+) f(x) = +")

print(f"\n:")
print(f"  -    ,     0")
print(f"  -   x=0    ( )")
print(f"  -  x3    +")
```

```python
# 2 

import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import minimize_scalar

def f(x):
    b = 2
    return np.sqrt(2/np.pi) *b**(3/2)*(x**2) * np.exp(-b*x**2/2)

x = np.linspace(0, 4, 1000)
y = f(x)

plt.figure(figsize=(12, 8))

plt.plot(x, y, 'b-', linewidth=2, label=r'$f(x) = \sqrt{\frac{2}{\pi}} x^2 e^{-x^2/2}$')

plt.axhline(y=0, color='r', linestyle='--', alpha=0.7, label=' y=0')

plt.plot(0, f(0), 'ro', markersize=8, label=' (0,0)')

x_extremum2 = np.sqrt(2)
y_extremum = f(np.sqrt(2))

plt.plot(x_extremum2, y_extremum, 'go', markersize=8)

def find_inflection_points():
    def second_derivative(x):
        return np.sqrt(2/np.pi) * np.exp(-x**2/2) * (x**4 - 5*x**2 + 2)
    
    inflection_points = []
    for guess in [ 0, 1, 2.5]:
        try:
            result = minimize_scalar(lambda x: abs(second_derivative(x)), 
                                   bounds=(guess-0.5, guess+0.5), 
                                   method='bounded')
            if result.success:
                root = result.x
                if abs(second_derivative(root)) < 1e-5 and root not in [r[0] for r in inflection_points]:
                    inflection_points.append((root, f(root)))
        except:
            pass
    
    return inflection_points

inflection_points = find_inflection_points()
for ip in inflection_points:
    plt.plot(ip[0], ip[1], 'mo', markersize=8, 
             label=f'  ({ip[0]:.2f}, {ip[1]:.2f})')

plt.xlabel('x')
plt.ylabel('f(x)')
plt.title(r'  $f(x) = \sqrt{\frac{2}{\pi}} x^2 e^{-x^2/2}$')
plt.grid(True, alpha=0.3)
plt.legend()
plt.ylim(-1, 1)
plt.axvline(x=np.sqrt(2), color='g', linestyle=':', alpha=0.5)
plt.show()

print("=" * 60)
print("  ")
print("=" * 60)

print(f"\n1.    :")
print(f"      OY: (0, 0)")
print(f"      OX: (0, 0)")

print(f"\n2. :")
print(f"     :")
print(f"     x = -2  {-np.sqrt(2):.3f}, y = {f(-np.sqrt(2)):.3f}")
print(f"     x = 2   {np.sqrt(2):.3f},  y = {f(np.sqrt(2)):.3f}")
print(f"     :")
print(f"     x = 0, y = 0")

print(f"\n3. :")
print(f"     : y = 0")
print(f"     : ")
print(f"     : ")

print(f"\n4.   ():")
for i, (x_ip, y_ip) in enumerate(inflection_points, 1):
    print(f"     {i}: x  {x_ip:.3f}, y  {y_ip:.3f}")

print(f"\n5.  :")
print(f"     : (-, +)")
print(f"     : [0, +)")
print(f"     : f(-x) = f(x)")
print(f"      x: 0")
```

```python
from numpy import*
from matplotlib.pyplot import*
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import minimize_scalar

def f(x):
    return np.exp(-x) * np.sin(2 * np.pi * x)

x_plot = np.linspace(0, 2, 1000)
y_plot = f(x_plot)

def f_negative(x):
    return -f(x)  

def f_positive(x):
    return f(x)  

extremum_points = []

initial_guesses = [0.1, 0.4, 0.8, 1.2, 1.6]

for guess in initial_guesses:
    result_max = minimize_scalar(f_negative, bounds=(0, 2), method='bounded')
    if result_max.success:
        x_max = result_max.x
        y_max = f(x_max)
        extremum_points.append(('max', x_max, y_max))
    
    result_min = minimize_scalar(f_positive, bounds=(0, 2), method='bounded')
    if result_min.success:
        x_min = result_min.x
        y_min = f(x_min)
        extremum_points.append(('min', x_min, y_min))

unique_extremums = {}
for ext_type, x_val, y_val in extremum_points:
    key = (ext_type, round(x_val, 3))
    if key not in unique_extremums or abs(y_val) > abs(unique_extremums[key][1]):
        unique_extremums[key] = (x_val, y_val)


maxima = [(x, y) for (typ, _), (x, y) in unique_extremums.items() if typ == 'max']
minima = [(x, y) for (typ, _), (x, y) in unique_extremums.items() if typ == 'min']


plt.figure(figsize=(12, 8))


plt.plot(x_plot, y_plot, 'b-', linewidth=2, label=r'$f(x) = e^{-x} \cdot \sin(2\pi x)$')

if maxima:
    max_x, max_y = zip(*maxima)
    plt.plot(max_x, max_y, 'ro', markersize=8, label=' ')
    for i, (x, y) in enumerate(maxima):
        plt.annotate(f'max({x:.3f}, {y:.3f})', 
                    xy=(x, y), xytext=(10, 10), 
                    textcoords='offset points', fontsize=10,
                    arrowprops=dict(arrowstyle='->', color='red'))

if minima:
    min_x, min_y = zip(*minima)
    plt.plot(min_x, min_y, 'go', markersize=8, label=' ')
    for i, (x, y) in enumerate(minima):
        plt.annotate(f'min({x:.3f}, {y:.3f})', 
                    xy=(x, y), xytext=(10, -20), 
                    textcoords='offset points', fontsize=10,
                    arrowprops=dict(arrowstyle='->', color='green'))


plt.xlabel('x', fontsize=12)
plt.ylabel('f(x)', fontsize=12)
plt.title('   $f(x) = e^{-x} \\cdot \\sin(2\\pi x)$  [0, 2]', fontsize=14)
plt.grid(True, alpha=0.3)
plt.legend()
plt.axhline(y=0, color='k', linestyle='-', alpha=0.3)
plt.axvline(x=0, color='k', linestyle='-', alpha=0.3)

roots = []
for i in range(len(x_plot)-1):
    if y_plot[i] * y_plot[i+1] <= 0:
        root = x_plot[i] - y_plot[i] * (x_plot[i+1] - x_plot[i]) / (y_plot[i+1] - y_plot[i])
        roots.append(root)
        plt.plot(root, 0, 'kx', markersize=6, alpha=0.7)

plt.tight_layout()
plt.show()

print("=" * 60)
print("  ")
print("=" * 60)
print(f": f(x) = e^(-x) * sin(2x)")
print(f": [0, 2]")
print()

print(":")
if maxima:
    for i, (x, y) in enumerate(maxima, 1):
        print(f"  {i}. x = {x:.4f}, f(x) = {y:.4f}")
else:
    print("    ")

print("\n:")
if minima:
    for i, (x, y) in enumerate(minima, 1):
        print(f"  {i}. x = {x:.4f}, f(x) = {y:.4f}")
else:
    print("    ")

print(f"\n (   OX):")
if roots:
    for i, root in enumerate(roots, 1):
        print(f"  {i}. x = {root:.4f}")
else:
    print("    ")

print(f"\n :")
print(f": f'(x) = -e^(-x)sin(2x) + 2e^(-x)cos(2x)")
print(f"  : -sin(2x) + 2cos(2x) = 0")
print(f": tan(2x) = 2")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import root_scalar

#  
def f(x):
    return np.sin(x) - x**2 * np.cos(x)

#  
def f_prime(x):
    return np.cos(x) - (2*x*np.cos(x) - x**2*(-np.sin(x)))
    #  : cos(x) - 2x*cos(x) - x^2*sin(x)

#  
def f_double_prime(x):
    return -np.sin(x) - (2*np.cos(x) - 2*x*np.sin(x) + 2*x*np.sin(x) + x**2*np.cos(x))
    #  : -sin(x) - 2cos(x) - x^2*cos(x) + 4x*sin(x)

#    x   
x_plot = np.linspace(-10, 10, 2000)
y_plot = f(x_plot)
y_double_prime = f_double_prime(x_plot)

#    (    )
inflection_points = []

for i in range(len(x_plot) - 1):
    if y_double_prime[i] * y_double_prime[i + 1] < 0:
        #      
        try:
            result = root_scalar(f_double_prime, bracket=[x_plot[i], x_plot[i + 1]], method='bisect')
            if result.converged:
                x_inflection = result.root
                y_inflection = f(x_inflection)
                inflection_points.append((x_inflection, y_inflection))
        except:
            continue

#   (  4 )
unique_inflection_points = []
tolerance = 1e-3
for point in inflection_points:
    is_duplicate = False
    for existing_point in unique_inflection_points:
        if abs(point[0] - existing_point[0]) < tolerance:
            is_duplicate = True
            break
    if not is_duplicate:
        unique_inflection_points.append(point)

#     x
unique_inflection_points.sort(key=lambda p: p[0])

#  
fig, (ax1, ax2, ax3) = plt.subplots(3, 1, figsize=(14, 12))

#  1:    
ax1.plot(x_plot, y_plot, 'b-', linewidth=2, label=r'$f(x) = \sin(x) - x^2\cos(x)$')
if unique_inflection_points:
    inf_x, inf_y = zip(*unique_inflection_points)
    ax1.plot(inf_x, inf_y, 'ro', markersize=8, label=' ')
    for i, (x, y) in enumerate(unique_inflection_points):
        ax1.annotate(f'({x:.2f}, {y:.2f})', 
                    xy=(x, y), xytext=(5, 10), 
                    textcoords='offset points', fontsize=8,
                    bbox=dict(boxstyle="round,pad=0.3", fc="yellow", alpha=0.7))
ax1.set_xlabel('x')
ax1.set_ylabel('f(x)')
ax1.set_title('   ')
ax1.grid(True, alpha=0.3)
ax1.legend()
ax1.axhline(y=0, color='k', linestyle='-', alpha=0.3)
ax1.axvline(x=0, color='k', linestyle='-', alpha=0.3)

#  2:  
ax2.plot(x_plot, [f_prime(x) for x in x_plot], 'g-', linewidth=2, label=r"$f'(x)$")
ax2.set_xlabel('x')
ax2.set_ylabel("f'(x)")
ax2.set_title(' ')
ax2.grid(True, alpha=0.3)
ax2.legend()
ax2.axhline(y=0, color='k', linestyle='-', alpha=0.3)

#  3:     
ax3.plot(x_plot, y_double_prime, 'r-', linewidth=2, label=r"$f''(x)$")
ax3.axhline(y=0, color='k', linestyle='-', alpha=0.5, linewidth=1)
if unique_inflection_points:
    ax3.plot(inf_x, [0]*len(inf_x), 'ro', markersize=6, label='  (f\'\'(x)=0)')
ax3.set_xlabel('x')
ax3.set_ylabel("f''(x)")
ax3.set_title('    ')
ax3.grid(True, alpha=0.3)
ax3.legend()

plt.tight_layout()
plt.show()

#  
print("=" * 70)
print("  ")
print("=" * 70)
print(f": f(x) = sin(x) - xcos(x)")
print(f": [-10, 10]")
print(f" : f'(x) = cos(x) - 2xcos(x) - xsin(x)")
print(f" : f''(x) = -sin(x) - 2cos(x) - xcos(x) + 4xsin(x)")
print()

print("  :")
if unique_inflection_points:
    for i, (x, y) in enumerate(unique_inflection_points, 1):
        #     
        left_val = f_double_prime(x - 0.01)
        right_val = f_double_prime(x + 0.01)
        sign_change = "+" if left_val < 0 and right_val > 0 else "-"
        
        print(f"  {i}. x = {x:8.4f}, f(x) = {y:8.4f}  {sign_change}")
        
        #  /
        if left_val < 0 and right_val > 0:
            print(f"          (f''   -  +)")
        else:
            print(f"          (f''   +  -)")
else:
    print("     ")

#   /
print(f"\n /:")
if unique_inflection_points:
    test_points = [-10] + [(unique_inflection_points[i][0] + unique_inflection_points[i+1][0])/2 
                          for i in range(len(unique_inflection_points)-1)] + [10]
    
    for i in range(len(test_points)):
        x_test = test_points[i]
        f_dp = f_double_prime(x_test)
        convexity = " (f'' > 0)" if f_dp > 0 else " (f'' < 0)"
        
        if i == 0:
            print(f"  (-, {unique_inflection_points[0][0]:.2f}): {convexity}")
        elif i == len(test_points) - 1:
            print(f"  ({unique_inflection_points[-1][0]:.2f}, +): {convexity}")
        else:
            print(f"  ({unique_inflection_points[i-1][0]:.2f}, {unique_inflection_points[i][0]:.2f}): {convexity}")
else:
    #    ,   
    f_dp_mid = f_double_prime(0)
    convexity = " (f'' > 0)" if f_dp_mid > 0 else " (f'' < 0)"
    print(f"      {convexity}")

#  
print(f"\n :")
print(f"    [-10, 10]: {len(unique_inflection_points)}")
print(f"  :     ")
```
