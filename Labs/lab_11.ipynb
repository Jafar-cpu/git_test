```python
#Упражнение 4

import matplotlib.pyplot as plt
import numpy as np
from sympy import*
fig = plt.figure(figsize=(8,8))
ax = fig.add_subplot(1, 1, 1, projection='3d')
#2,-3,5,10
d = Plane(Point(1,-1,-3),normal_vector=(2,-3,5))
print(d.equation())
f = lambda x,y:(-2*x+3*y-10)/5
xval = np.linspace(-4, 4, 100) # разбиение для x
yval = np.linspace(-4, 4, 100) # разбиение для y
x, y = np.meshgrid(xval, yval)
p = np.array([1,-1,-3]) # точка на плоскости
plt.plot(1,-1,-3,'ro')
v = np.array([2,-3,5]) # вектор нормали
ax.quiver(*p, *v, color='Blue') # строим вектор нормали
eps=0.5
ax.text(p[0] + eps, p[1] + eps,p[2]-eps,'$M(1,-1,-3)$')
ax.text(v[0] + eps, v[1] + eps,v[2]-eps,'$n(2,-3,4)$')
z = f(x,y) # вычисляем значения z в точках 
print(Point(0,0,0).distance(d))
surf = ax.plot_surface(x, y, z, rstride=10, cstride=10).set_alpha(0.5) # строим плоскость
ax.set_xlabel('ось X');ax.set_xlim(-4, 10)
ax.set_ylabel('ось Y');ax.set_ylim(-4, 10)
ax.set_zlabel('ось Z')
plt.title('Плоскость 2x - 3y + 5z + 10 = 0')
ax.view_init(8, -45) 
plt.show()
```

```python
# Упражнение 5

import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D


M = np.array([2, -3, -6])
N = np.array([-3, 1, 3])
a = np.array([-3, 2, -1])
b = np.array([0, 3, 1])
n = np.array([5, 3, -9])

lim = 10
x = np.linspace(-lim, lim, 20)
y = np.linspace(-lim, lim, 20)
X, Y = np.meshgrid(x, y)
Z = (5*X + 3*Y - 55) / 9

fig = plt.figure(figsize=(12, 10))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, alpha=0.5, color='lightblue', label='Плоскость')
ax.scatter(*M, color='red', s=100, label='M(2,-3,-6)')
ax.scatter(*N, color='blue', s=100, label='N(-3,1,3)')
ax.quiver(*M, *a, color='green', label='a(-3,2,-1)', linewidth=2, arrow_length_ratio=0.1)
ax.quiver(*M, *b, color='orange', label='b(0,3,1)', linewidth=2, arrow_length_ratio=0.1)


n_plot = n / 3  # Масштабируем для лучшего отображения
ax.quiver(*M, *n_plot, color='purple', label='n(5,3,-9)', linewidth=3, arrow_length_ratio=0.1)

t = -(5*N[0] + 3*N[1] - 9*N[2] - 55) / (5**2 + 3**2 + (-9)**2)
P = N + t * n  # Точка проекции
ax.plot([N[0], P[0]], [N[1], P[1]], [N[2], P[2]], 'k--', linewidth=2, label='Расстояние от N')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_xlim(-lim, lim)
ax.set_ylim(-lim, lim)
ax.set_zlim(-lim, lim)
distance = abs(5*N[0] + 3*N[1] - 9*N[2] - 55) / np.sqrt(5**2 + 3**2 + (-9)**2)
ax.set_title(f'Плоскость: $5x + 3y - 9z - 55 = 0$\nРасстояние от N до плоскости: {distance:.3f}', 
             fontsize=14)

ax.legend()
plt.tight_layout()
plt.show()

print(f"Нормальный вектор: n = (5, 3, -9)")
print(f"Расстояние от точки N до плоскости: {distance:.6f}")
```

```python
# Упражнение 6

import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D


M = np.array([1, 3, 0])
N = np.array([0, 4, -1])
n1 = np.array([1, -2, 3])  
n2 = np.array([-1, 10, 7])

lim = 8
x = np.linspace(-lim, lim, 20)
y = np.linspace(-lim, lim, 20)
X, Y = np.meshgrid(x, y)
Z1 = (-X + 2*Y + 10) / 3
Z2 = (X - 10*Y + 31) / 7

fig = plt.figure(figsize=(14, 10))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z1, alpha=0.5, color='lightblue', label='x - 2y + 3z - 10 = 0')
ax.plot_surface(X, Y, Z2, alpha=0.5, color='lightgreen', label='-x + 10y + 7z - 31 = 0')
ax.scatter(*M, color='red', s=100, label='M(1,3,0)')
ax.scatter(*N, color='blue', s=100, label='N(0,4,-1)')


origin = np.array([0, 0, 0])
ax.quiver(*origin, *n1, color='darkblue', label='n₁(1,-2,3)', linewidth=2, arrow_length_ratio=0.1)
ax.quiver(*origin, *n2, color='darkgreen', label='n₂(-1,10,7)', linewidth=2, arrow_length_ratio=0.1)


dot_product = np.dot(n1, n2)
perpendicular = np.isclose(dot_product, 0)

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_xlim(-lim, lim)
ax.set_ylim(-lim, lim)
ax.set_zlim(-lim, lim)

ax.set_title(f'Перпендикулярные плоскости\n'
             f'$x - 2y + 3z - 10 = 0$ и $-x + 10y + 7z - 31 = 0$\n'
             f'Скалярное произведение нормалей: {dot_product} → '
             f'{"Перпендикулярны" if perpendicular else "Не перпендикулярны"}', 
             fontsize=12)

ax.legend()
plt.tight_layout()
plt.show()

print(f"Нормальный вектор первой плоскости: n₁ = (1, -2, 3)")
print(f"Нормальный вектор второй плоскости: n₂ = (-1, 10, 7)")
print(f"Скалярное произведение: n₁·n₂ = {dot_product}")
print(f"Плоскости перпендикулярны: {perpendicular}")
```

```python
#Доп часть
import matplotlib.pyplot as plt
import numpy as np


lim = 10
x = np.linspace(-lim, lim, 20)
y = np.linspace(-lim, lim, 20)
X, Y = np.meshgrid(x, y)


Z1 = (2*X - Y - 3) / 12
Z2 = (3*X + Y - 2) / 7
Z3 = np.zeros_like(X)
Z4 = (-X - 2*Y - 1) / 5

# Построение
fig = plt.figure(figsize=(14, 10))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z1, alpha=0.3, color='lightblue', label='2x - y - 12z - 3 = 0')
ax.plot_surface(X, Y, Z2, alpha=0.3, color='lightgreen', label='3x + y - 7z - 2 = 0')
x_line = np.linspace(-lim, lim, 100)
y_line = 2*x_line - 12.5
z_line = np.zeros_like(x_line)
ax.plot(x_line, y_line, z_line, 'r-', linewidth=3, label='4x - 2y = 25')
ax.plot_surface(X, Y, Z4, alpha=0.5, color='orange', label='x + 2y + 5z + 1 = 0')

n1 = np.array([2, -1, -12])
n2 = np.array([3, 1, -7])
n3 = np.array([4, -2, 0])
n4 = np.array([1, 2, 5])

origin = np.array([0, 0, 0])
ax.quiver(*origin, *n1/3, color='blue', linewidth=2, arrow_length_ratio=0.1, label='n₁')
ax.quiver(*origin, *n2/3, color='green', linewidth=2, arrow_length_ratio=0.1, label='n₂')
ax.quiver(*origin, *n3/3, color='red', linewidth=2, arrow_length_ratio=0.1, label='n₃')
ax.quiver(*origin, *n4/3, color='orange', linewidth=2, arrow_length_ratio=0.1, label='n₄')

# Оформление
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_xlim(-lim, lim)
ax.set_ylim(-lim, lim)
ax.set_zlim(-lim, lim)

ax.set_title('Плоскость через линию пересечения, перпендикулярная заданной', fontsize=12)
ax.legend()

plt.tight_layout()
plt.show()

print("Уравнение искомой плоскости: x + 2y + 5z + 1 = 0")
print("Проверка перпендикулярности с плоскостью 4x - 2y = 25:")
print(f"n·n₃ = {np.dot(n4, n3)} → {'Перпендикулярны' if np.dot(n4, n3) == 0 else 'Не перпендикулярны'}")
```
