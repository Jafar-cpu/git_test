```python
#3.1

from numpy import*
import sympy as sp
A = matrix('3 -5;2 7')
B = matrix('13;81')
A_det = linalg.det(A)
X_m = matrix(A)
X_m[:, 0] = B
Y_m = matrix(A)
Y_m[:, 1] = B
x = linalg.det(X_m) / A_det
y = linalg.det(Y_m) / A_det
X = sp.Matrix([[x],[y]])
print(dot(A,X) - B)
print(x,y)
```
