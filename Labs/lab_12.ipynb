```python
import numpy as np

def distance_between_lines(A, s1, B, s2):
    """
    A, s1 — точка и направляющий вектор первой прямой
    B, s2 — точка и направляющий вектор второй прямой
    Возвращает расстояние по алгоритму из Упражнения 5.
    """
    A = np.array(A, dtype=float)
    s1 = np.array(s1, dtype=float)
    B = np.array(B, dtype=float)
    s2 = np.array(s2, dtype=float)
    
    # Шаг 1: плоскость P: s1·(r - A) = 0
    
    # Шаг 2: проекция L2 на P
    B0 = B
    B1 = B + s2
    
    # Проекция B0 на P
    # lambda0 = - s1·(B0 - A) / |s1|^2
    lambda0 = -np.dot(s1, B0 - A) / np.dot(s1, s1)
    B0_proj = B0 + lambda0 * s1
    
    # Проекция B1 на P
    lambda1 = -np.dot(s1, B1 - A) / np.dot(s1, s1)
    B1_proj = B1 + lambda1 * s1
    
    # Шаг 3: расстояние от A до прямой через B0_proj и B1_proj
    u = B1_proj - B0_proj
    AB0 = B0_proj - A

    cross = np.cross(AB0, u)
    d = np.linalg.norm(cross) / np.linalg.norm(u)
    return d

A = (9, 0, 2)
s1 = (-6, 2, 1)
B = (-5, -5, 1)
s2 = (3, 2, -2)

dist = distance_between_lines(A, s1, B, s2)
print(f"Расстояние между прямыми: {dist}")
print(f"Ожидаемый ответ: 7, получено: {dist:.6f}")
```
