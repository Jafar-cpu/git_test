```python
# 1 Задание 

from sympy import *
import matplotlib.pyplot as pt
import numpy as np 
from math import*
x = Symbol('x')
s = np.linspace(1,10)
t  = (np.e**(4*s) - 1)/(1 - np.e**(5*s))
t1 = (-s**2 - 3*s - 7)*np.e**(2*s)
t2 = np.sin(s)/(np.sqrt(s+1) - np.sqrt(2*s+1))

# Пределы  а,б,в
lim = limit(t , x,oo)
lim1 = limit(t1,x,oo)
lim2 = limit(t2,x,0)
print(lim,lim1,lim2)

# А
pt.subplot(1,3,1)
pt.plot(s,t,'b.',label='1 график')
pt.title('1 График')
pt.subplot(1,3,2)
# Б
pt.plot(s,t1,'b.',label='2 график')
pt.title('2 График')
# В
pt.subplot(1,3,3)
pt.plot(s,t2,'b.',label='2 график')
pt.title('3 График')
pt.tight_layout()
pt.show()
```
