```python
import numpy as np
import matplotlib.pyplot as plt

p = 0.039  
m = 0.3    
l = 1.0    

def M(x):
    """Масса участка нити длиной x"""
    return p * x + m * np.floor(x) + m


plt.figure(figsize=(10, 6))


for n in range(0, 6):

    x_left = n

    x_right = n + 1 - 1e-10 
    
    x_segment = np.linspace(x_left, x_right, 100)
    y_segment = M(x_segment)
    
    plt.plot(x_segment, y_segment, 'b-', linewidth=1.5)


for n in range(0, 7):
    y_value = M(n)
    plt.plot(n, y_value, 'bo', markersize=5, label=f'M({n})' if n == 0 else "")


for n in range(0, 7):
    plt.axvline(n, color='gray', linestyle=':', alpha=0.5)


plt.xlabel('Длина участка x (см)', fontsize=12)
plt.ylabel('Масса M(x) (г)', fontsize=12)
plt.title('График массы M(x) участка нити с бусинками', fontsize=14)
plt.grid(True, alpha=0.3)
plt.xlim(0, 6)
plt.ylim(0, M(6) + 0.1)


plt.legend(['M(x)', 'Значения в целых точках'])

plt.tight_layout()
plt.show()

print("Анализ разрывов:")
print("x\tM(x⁻)\tM(x)\tM(x⁺)\tСкачок")
for n in range(1, 7):
    M_left = p * n + m * (n - 1) + m     
    M_point = p * n + m * n + m          
    M_right = M_point                     
    jump = M_point - M_left
    
    print(f"{n}\t{M_left:.3f}\t{M_point:.3f}\t{M_right:.3f}\t{jump:.3f}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp

# Настройка для русского языка и улучшение качества графиков
plt.rcParams['font.size'] = 12
plt.rcParams['figure.figsize'] = [10, 6]

def plot_function_b():
    # Создаем символьную переменную и функцию
    x = sp.Symbol('x')
    f = x * sp.exp(-1/x)
    
    # Вычисляем асимптоту
    k = sp.limit(f/x, x, sp.oo)  # k = 1
    b = sp.limit(f - k*x, x, sp.oo)  # b = -1
    
    print(f"Функция: f(x) = x * e^(-1/x)")
    print(f"Наклонная асимптота: y = {float(k):.1f}x {float(b):+.1f}")
    
    # Преобразуем символьную функцию в числовую
    f_num = sp.lambdify(x, f, 'numpy')
    
    # Создаем массивы значений x
    # Для отрицательных x (исключая 0)
    x_neg = np.linspace(-10, -0.1, 1000)
    # Для положительных x (исключая 0)
    x_pos_small = np.linspace(0.1, 2, 500)
    x_pos_large = np.linspace(2, 20, 1000)
    
    # Вычисляем значения функции
    y_neg = f_num(x_neg)
    y_pos_small = f_num(x_pos_small)
    y_pos_large = f_num(x_pos_large)
    
    # Вычисляем значения асимптоты
    asymptote = k * x_pos_large + b
    
    # Создаем график
    plt.figure(figsize=(12, 8))
    
    # Рисуем функцию
    plt.plot(x_neg, y_neg, 'b-', linewidth=2, label='f(x) = x · e^(-1/x)')
    plt.plot(x_pos_small, y_pos_small, 'b-', linewidth=2)
    plt.plot(x_pos_large, y_pos_large, 'b-', linewidth=2)
    
    # Рисуем асимптоту
    plt.plot(x_pos_large, asymptote, 'r--', linewidth=2, 
             label=f'Асимптота: y = x - 1')
    
    # Добавляем вертикальную асимптоту в x = 0
    plt.axvline(x=0, color='green', linestyle=':', alpha=0.7, 
                label='Вертикальная асимптота: x = 0')
    
    # Настройка графика
    plt.xlabel('x', fontsize=14)
    plt.ylabel('f(x)', fontsize=14)
    plt.title('График функции f(x) = x · e^(-1/x) и ее асимптот', fontsize=16)
    plt.legend(fontsize=12)
    plt.grid(True, alpha=0.3)
    plt.axhline(y=0, color='k', linewidth=0.5)
    plt.axvline(x=0, color='k', linewidth=0.5)
    
    # Устанавливаем пределы осей для лучшего обзора
    plt.xlim(-5, 15)
    plt.ylim(-10, 15)
    
    # Добавляем дополнительные элементы для наглядности
    plt.text(12, 12, 'f(x) = x · e^(-1/x)', fontsize=12, color='blue')
    plt.text(12, 9, 'y = x - 1 (асимптота)', fontsize=12, color='red')
    
    plt.tight_layout()
    plt.show()

def detailed_analysis():
    """Детальный анализ поведения функции"""
    print("\n=== Детальный анализ функции ===")
    
    x = sp.Symbol('x')
    f = x * sp.exp(-1/x)
    
    # Пределы в особых точках
    print("Пределы:")
    print(f"lim(x→0+) f(x) = {sp.limit(f, x, 0, '+')}")
    print(f"lim(x→0-) f(x) = {sp.limit(f, x, 0, '-')}")
    print(f"lim(x→∞) f(x) = {sp.limit(f, x, sp.oo)}")
    print(f"lim(x→-∞) f(x) = {sp.limit(f, x, -sp.oo)}")
    
    # Производная для анализа монотонности
    f_prime = sp.diff(f, x)
    print(f"\nПроизводная: f'(x) = {sp.simplify(f_prime)}")
    
    # Критические точки
    critical_points = sp.solve(f_prime, x)
    print(f"Критические точки: {critical_points}")

# Запускаем построение графика
plot_function_b()
```
